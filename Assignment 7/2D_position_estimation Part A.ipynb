By definition, the constant-velocity kinematic equations over a discrete time increment $\Delta t = t_k - t_{k-1}$ dictate that the new position is the old position plus the displacement due to velocity, and velocity remains constant.

We write the algebraic system of equations for the state components at time $k$ given time $k-1$ (omitting noise for matrix isolation):

$$\begin{aligned}
p_x(k) &= 1 \cdot p_x(k-1) + 0 \cdot p_y(k-1) + \Delta t \cdot v_x(k-1) + 0 \cdot v_y(k-1) \\
p_y(k) &= 0 \cdot p_x(k-1) + 1 \cdot p_y(k-1) + 0 \cdot v_x(k-1) + \Delta t \cdot v_y(k-1) \\
v_x(k) &= 0 \cdot p_x(k-1) + 0 \cdot p_y(k-1) + 1 \cdot v_x(k-1) + 0 \cdot v_y(k-1) \\
v_y(k) &= 0 \cdot p_x(k-1) + 0 \cdot p_y(k-1) + 0 \cdot v_x(k-1) + 1 \cdot v_y(k-1)
\end{aligned}$$

We map these coefficients directly into a system of vector-matrix multiplications of the form $x^-_k = A x^+_{k-1}$:

$$\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix}
=
\begin{bmatrix}
\mathbf{1} & \mathbf{0} & \mathbf{\Delta t} & \mathbf{0}\\
\mathbf{0} & \mathbf{1} & \mathbf{0} & \mathbf{\Delta t}\\
\mathbf{0} & \mathbf{0} & \mathbf{1} & \mathbf{0}\\
\mathbf{0} & \mathbf{0} & \mathbf{0} & \mathbf{1}
\end{bmatrix}
\begin{bmatrix}
p_x(k-1)\\
p_y(k-1)\\
v_x(k-1)\\
v_y(k-1)
\end{bmatrix}$$

Extracting the operator matrix yields:

$$A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix}$$

$$\text{Q.E.D.}$$

Let the process noise vector be defined by random accelerations $w_{k-1} = \begin{bmatrix} a_x(k-1) \\ a_y(k-1) \end{bmatrix}$.

Integrating a constant acceleration twice with respect to time yields the position displacement $\int \int a \,dt^2 = \frac{1}{2}a\Delta t^2$, and integrating once yields the velocity change $\int a \,dt = a\Delta t$.

Expressing the total algebraic contribution of $w_{k-1}$ on each state component:

$$\begin{aligned}
\Delta p_x &= \frac{1}{2}\Delta t^2 \cdot a_x(k-1) + 0 \cdot a_y(k-1) \\
\Delta p_y &= 0 \cdot a_x(k-1) + \frac{1}{2}\Delta t^2 \cdot a_y(k-1) \\
\Delta v_x &= \Delta t \cdot a_x(k-1) + 0 \cdot a_y(k-1) \\
\Delta v_y &= 0 \cdot a_x(k-1) + \Delta t \cdot a_y(k-1)
\end{aligned}$$

Putting this system of equations into the vector format $\Delta x = G w_{k-1}$:

$$\begin{bmatrix}
\Delta p_x\\
\Delta p_y\\
\Delta v_x\\
\Delta v_y
\end{bmatrix}
=
\begin{bmatrix}
\mathbf{\frac{1}{2}\Delta t^2} & \mathbf{0}\\
\mathbf{0} & \mathbf{\frac{1}{2}\Delta t^2}\\
\mathbf{\Delta t} & \mathbf{0}\\
\mathbf{0} & \mathbf{\Delta t}
\end{bmatrix}
\begin{bmatrix}
a_x(k-1)\\
a_y(k-1)
\end{bmatrix}$$

Extracting the mapping matrix yields:

$$G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}$$

The problem states that the sensor only tracks position elements, meaning the observation vector $y_k$ maps directly out of the state vector components.

Setting up the coordinate equations explicitly:


$$\begin{aligned}
p_x^{\mathrm{meas}}(k) &= 1 \cdot p_x(k) + 0 \cdot p_y(k) + 0 \cdot v_x(k) + 0 \cdot v_y(k) \\
p_y^{\mathrm{meas}}(k) &= 0 \cdot p_x(k) + 1 \cdot p_y(k) + 0 \cdot v_x(k) + 0 \cdot v_y(k)
\end{aligned}$$

Writing this in the linear observation format $y_k = Hx_k$:

$$\begin{bmatrix}
p_x^{\mathrm{meas}}(k)\\
p_y^{\mathrm{meas}}(k)
\end{bmatrix}
=
\begin{bmatrix}
\mathbf{1} & \mathbf{0} & \mathbf{0} & \mathbf{0}\\
\mathbf{0} & \mathbf{1} & \mathbf{0} & \mathbf{0}
\end{bmatrix}
\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix}$$

Isolating the linear operator matrix yields the final measurement matrix:

$$H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix}$$

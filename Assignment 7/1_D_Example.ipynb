### **Question 1 Solution (1-D Example)**

Given the scalar system model:

$$x^-_k = a x^+_{k-1} + w_{k-1}$$

where $x^+_{k-1} \sim \mathcal{N}(m_{k-1}, P_{k-1})$, $w_{k-1} \sim \mathcal{N}(0, q)$, and $x^+_{k-1}, w_{k-1}$ are independent scalar random variables.

---

#### **1. Evaluation of Predicted Mean ($m_k^-$)**

$$m_k^- = E[x^-_k]$$

$$m_k^- = E[a x^+_{k-1} + w_{k-1}]$$

By linearity of expectation:

$$m_k^- = a E[x^+_{k-1}] + E[w_{k-1}]$$

Substituting $E[x^+_{k-1}] = m_{k-1}$ and $E[w_{k-1}] = 0$:

$$\mathbf{m_k^- = a m_{k-1}}$$

---

#### **2. Evaluation of Predicted Variance ($P_k^-$)**

$$P_k^- = \text{Var}(x^-_k)$$

$$P_k^- = \text{Var}(a x^+_{k-1} + w_{k-1})$$

Since $x^+_{k-1}$ and $w_{k-1}$ are independent, the variance of their sum equals the sum of their variances:

$$P_k^- = \text{Var}(a x^+_{k-1}) + \text{Var}(w_{k-1})$$

Using the variance property $\text{Var}(c X) = c^2 \text{Var}(X)$ for a constant $c$:

$$P_k^- = a^2 \text{Var}(x^+_{k-1}) + \text{Var}(w_{k-1})$$

Substituting $\text{Var}(x^+_{k-1}) = P_{k-1}$ and $\text{Var}(w_{k-1}) = q$:

$$\mathbf{P_k^- = a^2 P_{k-1} + q}$$

---

#### **Conclusion**

$$\begin{aligned}
m_k^- &= a m_{k-1} \\
P_k^- &= a^2 P_{k-1} + q
\end{aligned}$$



### **Question 2 Solution (1-D Example)**

Given the scalar measurement model:

$$y_k = h x_k^- + z_k$$

where $x_k^- \sim \mathcal{N}(m_k^-, P_k^-)$, $z_k \sim \mathcal{N}(0, r)$, and the measurement innovation is defined as $v_k = y^{\mathrm{obs}}_k - h m_k^-$.

From the scalar simplification of the general Kalman filtering conditioning equations, the innovation variance $S_k$ is:

$$S_k = \text{Var}(y_k^-) = h^2 P_k^- + r$$

The scalar Kalman gain $K_k$ is defined as:

$$K_k = \frac{\text{Cov}(x_k^-, y_k^-)}{\text{Var}(y_k^-)} = \frac{P_k^- h}{S_k}$$

---

#### **1. Evaluation of Updated Mean ($m_k$)**

Applying the scalar Gaussian conditioning formula for the mean:

$$m_k = m_k^- + K_k (y^{\mathrm{obs}}_k - h m_k^-)$$

Substituting $v_k = y^{\mathrm{obs}}_k - h m_k^-$:

$$m_k = m_k^- + K_k v_k$$

Substituting the scalar expression for $K_k = \frac{P_k^- h}{S_k}$:

$$\mathbf{m_k = m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h m_k^-\bigr)}$$

---

#### **2. Evaluation of Updated Variance ($P_k$)**

Applying the scalar Gaussian conditioning formula for the variance:

$$P_k = P_k^- - K_k \cdot \text{Cov}(y_k^-, x_k^-)$$

Since $\text{Cov}(y_k^-, x_k^-) = h P_k^-$:

$$P_k = P_k^- - K_k (h P_k^-)$$

Factoring out $P_k^-$:

$$P_k = (1 - K_k h) P_k^-$$

Substituting the scalar expression for $K_k = \frac{P_k^- h}{S_k}$:

$$\mathbf{P_k = \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^-}$$

---

#### **Conclusion**

$$\begin{aligned}
m_k &= m_k^- + K_k v_k = m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h m_k^-\bigr) \\
P_k &= (1 - K_k h) P_k^- = \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^-
\end{aligned}$$


### **Question 3 Solution (1-D Example)**

Given the scalar measurement model:

$$y^-_k = h x^-_k + z_k$$

where $x^-_k \mid Y_{k-1} \sim \mathcal{N}(m_k^-, P_k^-)$, $z_k \sim \mathcal{N}(0, r)$, and $x^-_k$ and $z_k$ are independent scalar random variables.

Since $y^-_k$ is a linear combination of independent Gaussian random variables, its conditional distribution $p(y^-_k \mid Y_{k-1})$ is also Gaussian:

$$y^-_k \mid Y_{k-1} \sim \mathcal{N}(\mu_{y|Y}, \sigma^2_{y|Y})$$

---

#### **1. Evaluation of Conditional Mean ($\mu_{y|Y}$)**

$$\mu_{y|Y} = E[y^-_k \mid Y_{k-1}]$$

$$\mu_{y|Y} = E[h x^-_k + z_k \mid Y_{k-1}]$$

By linearity of expectation:

$$\mu_{y|Y} = h E[x^-_k \mid Y_{k-1}] + E[z_k \mid Y_{k-1}]$$

Substituting $E[x^-_k \mid Y_{k-1}] = m_k^-$ and $E[z_k \mid Y_{k-1}] = E[z_k] = 0$:

$$\mathbf{\mu_{y|Y} = h m_k^-}$$

---

#### **2. Evaluation of Conditional Variance ($\sigma^2_{y|Y}$)**

$$\sigma^2_{y|Y} = \text{Var}(y^-_k \mid Y_{k-1})$$

$$\sigma^2_{y|Y} = \text{Var}(h x^-_k + z_k \mid Y_{k-1})$$

Since $x^-_k$ and $z_k$ are conditionally independent, the variance of their sum is the sum of their variances:

$$\sigma^2_{y|Y} = \text{Var}(h x^-_k \mid Y_{k-1}) + \text{Var}(z_k \mid Y_{k-1})$$

Using the variance property $\text{Var}(c X) = c^2 \text{Var}(X)$ for a constant $c$:

$$\sigma^2_{y|Y} = h^2 \text{Var}(x^-_k \mid Y_{k-1}) + \text{Var}(z_k)$$

Substituting $\text{Var}(x^-_k \mid Y_{k-1}) = P_k^-$ and $\text{Var}(z_k) = r$:

$$\mathbf{\sigma^2_{y|Y} = h^2 P_k^- + r}$$

---

#### **Conclusion**

$$p(y^-_k \mid Y_{k-1}) = \mathcal{N}\bigl(h m_k^-,\; h^2 P_k^- + r\bigr)$$



### **Question 4 Solution (1-D Example)**

Given the scalar measurement model at a future step or evaluating the predictive distribution conditioned on the current filtered history $Y_k$:

$$y^-_k = h x_k + z_k$$

where $x_k \mid Y_k \sim \mathcal{N}(m_k, P_k)$, $z_k \sim \mathcal{N}(0, r)$, and $x_k$ and $z_k$ are independent scalar random variables.

Since $y^-_k$ is a linear combination of independent Gaussian random variables, its conditional distribution $p(y^-_k \mid Y_k)$ is also Gaussian:

$$y^-_k \mid Y_k \sim \mathcal{N}(\mu_{y|Y_k}, \sigma^2_{y|Y_k})$$

---

#### **1. Evaluation of Conditional Mean ($\mu_{y|Y_k}$)**

$$\mu_{y|Y_k} = E[y^-_k \mid Y_k]$$

$$\mu_{y|Y_k} = E[h x_k + z_k \mid Y_k]$$

By linearity of expectation:

$$\mu_{y|Y_k} = h E[x_k \mid Y_k] + E[z_k \mid Y_k]$$

Substituting $E[x_k \mid Y_k] = m_k$ and $E[z_k \mid Y_k] = E[z_k] = 0$:

$$\mathbf{\mu_{y|Y_k} = h m_k}$$

---

#### **2. Evaluation of Conditional Variance ($\sigma^2_{y|Y_k}$)**

$$\sigma^2_{y|Y_k} = \text{Var}(y^-_k \mid Y_k)$$

$$\sigma^2_{y|Y_k} = \text{Var}(h x_k + z_k \mid Y_k)$$

Since $x_k$ and $z_k$ are conditionally independent given $Y_k$, the variance of their sum is the sum of their variances:

$$\sigma^2_{y|Y_k} = \text{Var}(h x_k \mid Y_k) + \text{Var}(z_k \mid Y_k)$$

Using the variance property $\text{Var}(c X) = c^2 \text{Var}(X)$ for a constant $c$:

$$\sigma^2_{y|Y_k} = h^2 \text{Var}(x_k \mid Y_k) + \text{Var}(z_k)$$

Substituting $\text{Var}(x_k \mid Y_k) = P_k$ and $\text{Var}(z_k) = r$:

$$\mathbf{\sigma^2_{y|Y_k} = h^2 P_k + r}$$

---

#### **Conclusion**

$$p(y^-_k \mid Y_k) = \mathcal{N}\bigl(h m_k,\; h^2 P_k + r\bigr)$$

$$\tag*{$\blacksquare$}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from IPython.display import clear_output, display
import time

# Set random seed for reproducibility
np.random.seed(42)

# --- 1. MODEL PARAMETERS ---
a = 0.95  # Transition factor
q = 1.0   # Process noise variance
h = 1.0   # Measurement sensitivity
r = 0.8   # Measurement noise variance

# Initial Prior Distribution
m_current = 0.0
P_current = 3.0

# --- 2. GENERATE SIMULATED DATA ---
T = 5
true_x = []
obs_y = []
x_true = 2.0  # Initial true state

for k in range(T):
    x_true = a * x_true + np.random.normal(0, np.sqrt(q))
    y_val = h * x_true + np.random.normal(0, np.sqrt(r))
    true_x.append(x_true)
    obs_y.append(y_val)

# --- 3. NOTEBOOK-COMPATIBLE ANIMATION LOOP ---
x_axis = np.linspace(-7, 7, 500)

for k in range(T):
    # ==================================================
    # STEP A: PREDICTION PHASE
    # ==================================================
    m_pred = a * m_current
    P_pred = a**2 * P_current + q

    fig, ax = plt.subplots(figsize=(9, 5))
    ax.plot(x_axis, norm.pdf(x_axis, m_current, np.sqrt(P_current)), label=f'Posterior $x_{{{k}}}^+$', color='blue', lw=2)
    ax.plot(x_axis, norm.pdf(x_axis, m_pred, np.sqrt(P_pred)), label=f'Predicted (Prior) $x_{{{k+1}}}^-$', color='orange', lw=2, linestyle='--')
    ax.axvline(true_x[k], color='green', linestyle=':', label='True State $x_k$', lw=2)

    ax.set_title(f'Time Step {k+1}: Prediction Phase (Uncertainty Grows)')
    ax.set_xlabel('State $x$')
    ax.set_ylabel('Probability Density')
    ax.set_ylim(0, 0.7)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)

    # Force notebook to clear previous output and show current frame
    clear_output(wait=True)
    display(fig)
    plt.close(fig)
    time.sleep(2.0)  # Pause to view prediction expansion

    # ==================================================
    # STEP B: UPDATE PHASE
    # ==================================================
    y_obs = obs_y[k]
    S = h**2 * P_pred + r
    K = (P_pred * h) / S
    m_current = m_pred + K * (y_obs - h * m_pred)
    P_current = (1 - K * h) * P_pred

    fig, ax = plt.subplots(figsize=(9, 5))
    ax.plot(x_axis, norm.pdf(x_axis, m_pred, np.sqrt(P_pred)), label=f'Predicted (Prior) $x_{{{k+1}}}^-$', color='orange', lw=2, linestyle='--')
    ax.axvline(y_obs / h, color='red', linestyle='-.', label=f'Measurement $y_{{{k+1}}}/h$', lw=2)
    ax.plot(x_axis, norm.pdf(x_axis, m_current, np.sqrt(P_current)), label=f'Updated Posterior $x_{{{k+1}}}^+$', color='purple', lw=2)
    ax.axvline(true_x[k], color='green', linestyle=':', label='True State $x_k$', lw=2)

    ax.set_title(f'Time Step {k+1}: Update Phase (Uncertainty Shrinks)')
    ax.set_xlabel('State $x$')
    ax.set_ylabel('Probability Density')
    ax.set_ylim(0, 0.7)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)

    # Force notebook to clear previous output and show current frame
    clear_output(wait=True)
    display(fig)
    plt.close(fig)
    time.sleep(2.5)  # Pause to view update contraction
```

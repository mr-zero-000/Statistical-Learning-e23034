# **Question 01**
### Context & Assumptions

We are given the state transition equation:


$$x_k^- = A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1}$$

Where:

* The prior state is normally distributed: $x_{k-1}^+ \sim \mathcal{N}(m_{k-1}, P_{k-1})$
* The process noise is normally distributed: $w_{k-1} \sim \mathcal{N}(0, \Sigma_p)$
* $x_{k-1}^+$ and $w_{k-1}$ are assumed to be statistically independent.

Because a linear combination of independent Gaussian random variables is also Gaussian, the predicted state $x_k^-$ will follow a normal distribution:


$$x_k^- \sim \mathcal{N}(m_k^-, P_k^-)$$

To completely define this distribution, we need to derive its mean ($m_k^-$) and covariance ($P_k^-$).

---

### 1. Derivation of the Mean ($m_k^-$)

The expected value (mean) of the predicted state is found by applying the linear property of the expectation operator $E[\cdot]$:

$$m_k^- = E[x_k^-]$$

Substitute the state equation into the expectation:


$$m_k^- = E[A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1}]$$

$$m_k^- = A_{k-1} E[x_{k-1}^+] + G_{k-1} E[w_{k-1}]$$

Since $E[x_{k-1}^+] = m_{k-1}$ and $E[w_{k-1}] = 0$:


$$m_k^- = A_{k-1} m_{k-1} + G_{k-1}(0)$$

$$\mathbf{m_k^- = A_{k-1} m_{k-1}}$$

---

### 2. Derivation of the Covariance ($P_k^-$)

The covariance of the predicted state is defined as:


$$P_k^- = E[(x_k^- - m_k^-)(x_k^- - m_k^-)^T]$$

First, let's substitute the expressions for $x_k^-$ and $m_k^-$ into the error term $(x_k^- - m_k^-)$:


$$x_k^- - m_k^- = (A_{k-1} x_{k-1}^+ + G_{k-1} w_{k-1}) - A_{k-1} m_{k-1}$$

Group the terms involving the state vector:


$$x_k^- - m_k^- = A_{k-1}(x_{k-1}^+ - m_{k-1}) + G_{k-1} w_{k-1}$$

Now, substitute this error term back into the covariance formula:


$$P_k^- = E\left[ \Big( A_{k-1}(x_{k-1}^+ - m_{k-1}) + G_{k-1} w_{k-1} \Big) \Big( A_{k-1}(x_{k-1}^+ - m_{k-1}) + G_{k-1} w_{k-1} \Big)^T \right]$$

Expanding the outer product inside the expectation using $(X+Y)^T = X^T + Y^T$:


$$P_k^- = E\left[ \Big( A_{k-1}(x_{k-1}^+ - m_{k-1}) + G_{k-1} w_{k-1} \Big) \Big( (x_{k-1}^+ - m_{k-1})^T A_{k-1}^T + w_{k-1}^T G_{k-1}^T \Big) \right]$$

Distributing the terms yields four separate components:


$$\begin{aligned}
P_k^- = \, &E\left[ A_{k-1}(x_{k-1}^+ - m_{k-1})(x_{k-1}^+ - m_{k-1})^T A_{k-1}^T \right] \\
&+ E\left[ A_{k-1}(x_{k-1}^+ - m_{k-1})w_{k-1}^T G_{k-1}^T \right] \\
&+ E\left[ G_{k-1} w_{k-1}(x_{k-1}^+ - m_{k-1})^T A_{k-1}^T \right] \\
&+ E\left[ G_{k-1} w_{k-1} w_{k-1}^T G_{k-1}^T \right]
\end{aligned}$$

Using the linearity of expectation, we can bring the deterministic matrices ($A_{k-1}, G_{k-1}$) outside the expectation operator:


$$\begin{aligned}
P_k^- = \, &A_{k-1} E\left[ (x_{k-1}^+ - m_{k-1})(x_{k-1}^+ - m_{k-1})^T \right] A_{k-1}^T \\
&+ A_{k-1} E\left[ (x_{k-1}^+ - m_{k-1})w_{k-1}^T \right] G_{k-1}^T \\
&+ G_{k-1} E\left[ w_{k-1}(x_{k-1}^+ - m_{k-1})^T \right] A_{k-1}^T \\
&+ G_{k-1} E\left[ w_{k-1} w_{k-1}^T \right] G_{k-1}^T
\end{aligned}$$

### Simplifying the Terms:

1. **First term:** By definition, $E[(x_{k-1}^+ - m_{k-1})(x_{k-1}^+ - m_{k-1})^T] = P_{k-1}$.
2. **Second & Third terms (Cross-covariance):** Because the state $x_{k-1}^+$ and the process noise $w_{k-1}$ are independent, their cross-covariance is zero:

$$E[(x_{k-1}^+ - m_{k-1})w_{k-1}^T] = 0$$


3. **Fourth term:** By definition, the covariance of the noise is $E[w_{k-1} w_{k-1}^T] = \Sigma_p$.

Substituting these back gives:


$$P_k^- = A_{k-1} P_{k-1} A_{k-1}^T + 0 + 0 + G_{k-1} \Sigma_p G_{k-1}^T$$

$$\mathbf{P_k^- = A_{k-1} P_{k-1} A_{k-1}^T + G_{k-1} \Sigma_p G_{k-1}^T}$$

---

### Conclusion

Since both properties match perfectly, the predicted state distribution is indeed:


$$x_k^- \sim \mathcal{N}(m_k^-, P_k^-)$$

### **Question 2 Solution**

Given the measurement model:


$$y_k = H_k x_k^- + v_k$$

where $x_k^- \sim \mathcal{N}(m_k^-, P_k^-)$, $v_k \sim \mathcal{N}(0, \Sigma_m)$, and $x_k^-, v_k$ are statistically independent.

Since $y_k$ is a linear combination of independent Gaussian random vectors, $y_k^-$ is also Gaussian distributed:


$$y_k^- \sim \mathcal{N}(\mu_{y}, \Sigma_{y})$$

---

#### **1. Evaluation of Mean ($\mu_{y}$)**

$$\mu_{y} = E[y_k]$$

$$\mu_{y} = E[H_k x_k^- + v_k]$$

By linearity of expectation:


$$\mu_{y} = H_k E[x_k^-] + E[v_k]$$

Substituting $E[x_k^-] = m_k^-$ and $E[v_k] = 0$:


$$\mu_{y} = H_k m_k^-$$

---

#### **2. Evaluation of Covariance ($\Sigma_{y}$)**

$$\Sigma_{y} = E[(y_k - \mu_{y})(y_k - \mu_{y})^T]$$

Substituting $y_k = H_k x_k^- + v_k$ and $\mu_{y} = H_k m_k^-$:


$$y_k - \mu_{y} = (H_k x_k^- + v_k) - H_k m_k^-$$

$$y_k - \mu_{y} = H_k(x_k^- - m_k^-) + v_k$$

Now expanding the covariance:


$$\Sigma_{y} = E\left[ \Big( H_k(x_k^- - m_k^-) + v_k \Big) \Big( H_k(x_k^- - m_k^-) + v_k \Big)^T \right]$$

$$\Sigma_{y} = E\left[ \Big( H_k(x_k^- - m_k^-) + v_k \Big) \Big( (x_k^- - m_k^-)^T H_k^T + v_k^T \Big) \right]$$

Expanding the outer product:


$$\begin{aligned}
\Sigma_{y} = \, &E\left[ H_k(x_k^- - m_k^-)(x_k^- - m_k^-)^T H_k^T \right] + E\left[ H_k(x_k^- - m_k^-)v_k^T \right] \\
&+ E\left[ v_k(x_k^- - m_k^-)^T H_k^T \right] + E\left[ v_k v_k^T \right]
\end{aligned}$$

Factoring out deterministic matrices:


$$\begin{aligned}
\Sigma_{y} = \, &H_k E\left[ (x_k^- - m_k^-)(x_k^- - m_k^-)^T \right] H_k^T + H_k E\left[ (x_k^- - m_k^-)v_k^T \right] \\
&+ E\left[ v_k(x_k^- - m_k^-)^T \right] H_k^T + E\left[ v_k v_k^T \right]
\end{aligned}$$

Since $x_k^-$ and $v_k$ are independent, their cross-covariance is zero:


$$E\left[ (x_k^- - m_k^-)v_k^T \right] = 0 \quad \text{and} \quad E\left[ v_k(x_k^- - m_k^-)^T \right] = 0$$

Using known definitions:


$$E\left[ (x_k^- - m_k^-)(x_k^- - m_k^-)^T \right] = P_k^-$$

$$E\left[ v_k v_k^T \right] = \Sigma_m$$

Substituting these yields:


$$\Sigma_{y} = H_k P_k^- H_k^T + 0 + 0 + \Sigma_m$$

$$\Sigma_{y} = H_k P_k^- H_k^T + \Sigma_m$$

---

#### **Conclusion**

$$y_k^- \sim \mathcal{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m)$$



### **Question 3 Solution**

Let the joint random vector be defined as:


$$z_k = \begin{bmatrix} x_k^- \\ y_k^- \end{bmatrix}$$

Since $x_k^-$ is Gaussian and $y_k^- = H_k x_k^- + v_k$ is a linear transformation of Gaussian variables, the joint vector $z_k$ is joint-Gaussian distributed:


$$z_k \sim \mathcal{N}(\mu_z, \Sigma_z)$$

---

#### **1. Evaluation of Joint Mean ($\mu_z$)**

$$\mu_z = E[z_k] = E\begin{bmatrix} x_k^- \\ y_k^- \end{bmatrix} = \begin{bmatrix} E[x_k^-] \\ E[y_k^-] \end{bmatrix}$$

From the properties of the prior distributions and previous results:


$$E[x_k^-] = m_k^-$$

$$E[y_k^-] = H_k m_k^-$$

Thus:


$$\mu_z = \begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix}$$

---

#### **2. Evaluation of Joint Covariance Matrix ($\Sigma_z$)**

The joint covariance matrix is structured as:


$$\Sigma_z = \begin{bmatrix} \text{Cov}(x_k^-, x_k^-) & \text{Cov}(x_k^-, y_k^-) \\ \text{Cov}(y_k^-, x_k^-) & \text{Cov}(y_k^-, y_k^-) \end{bmatrix}$$

From previous results, the diagonal blocks are known:


$$\text{Cov}(x_k^-, x_k^-) = P_k^-$$

$$\text{Cov}(y_k^-, y_k^-) = H_k P_k^- H_k^T + \Sigma_m$$

We evaluate the cross-covariance term $\text{Cov}(x_k^-, y_k^-)$:


$$\text{Cov}(x_k^-, y_k^-) = E\left[ (x_k^- - m_k^-)(y_k^- - H_k m_k^-)^T \right]$$

Substituting $y_k^- - H_k m_k^- = H_k(x_k^- - m_k^-) + v_k$:


$$\text{Cov}(x_k^-, y_k^-) = E\left[ (x_k^- - m_k^-) \Big( H_k(x_k^- - m_k^-) + v_k \Big)^T \right]$$

$$\text{Cov}(x_k^-, y_k^-) = E\left[ (x_k^- - m_k^-) \Big( (x_k^- - m_k^-)^T H_k^T + v_k^T \Big) \right]$$

Expanding the expectation:


$$\text{Cov}(x_k^-, y_k^-) = E\left[ (x_k^- - m_k^-)(x_k^- - m_k^-)^T \right] H_k^T + E\left[ (x_k^- - m_k^-)v_k^T \right]$$

Since $x_k^-$ and $v_k$ are independent, $E\left[ (x_k^- - m_k^-)v_k^T \right] = 0$. Using $E\left[ (x_k^- - m_k^-)(x_k^- - m_k^-)^T \right] = P_k^-$:


$$\text{Cov}(x_k^-, y_k^-) = P_k^- H_k^T$$

Since $\text{Cov}(y_k^-, x_k^-) = \Big( \text{Cov}(x_k^-, y_k^-) \Big)^T$:


$$\text{Cov}(y_k^-, x_k^-) = (P_k^- H_k^T)^T = H_k P_k^-$$

Assembling the blocks into $\Sigma_z$:


$$\Sigma_z = \begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}$$

---

#### **Conclusion**

$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathcal{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right)$$



### **Question 4 Solution**

To find the conditional distribution of $x_k^-$ given $y_k^- = y^{\mathrm{obs}}_k$, we apply the standard Gaussian conditioning theorem to the joint distribution derived in Question 3.

For a joint Gaussian distribution:


$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left( \begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix} \right)$$

The conditional distribution $(X \mid Y = y)$ is also Gaussian, $\mathcal{N}(\mu_{X|Y}, \Sigma_{X|Y})$, where:


$$\mu_{X|Y} = \mu_X + \Sigma_{XY} \Sigma_{YY}^{-1} (y - \mu_Y)$$

$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY} \Sigma_{YY}^{-1} \Sigma_{YX}$$

---

#### **1. Mapping Joint Distribution Terms**

From Question 3, we identify:


$$\mu_X = m_k^-, \quad \mu_Y = H_k m_k^-, \quad y = y^{\mathrm{obs}}_k$$

$$\Sigma_{XX} = P_k^-, \quad \Sigma_{XY} = P_k^- H_k^T, \quad \Sigma_{YX} = H_k P_k^-, \quad \Sigma_{YY} = H_k P_k^- H_k^T + \Sigma_m$$

---

#### **2. Evaluation of Updated Mean ($m_k$)**

Substituting the mapped components into the conditional mean formula:


$$m_k = m_k^- + \Big( P_k^- H_k^T \Big) \Big( H_k P_k^- H_k^T + \Sigma_m \Big)^{-1} (y^{\mathrm{obs}}_k - H_k m_k^-)$$

Defining the Kalman Gain matrix $K_k \triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$:


$$m_k = m_k^- + K_k (y^{\mathrm{obs}}_k - H_k m_k^-)$$

---

#### **3. Evaluation of Updated Covariance ($P_k$)**

Substituting the mapped components into the conditional covariance formula:


$$P_k = P_k^- - \Big( P_k^- H_k^T \Big) \Big( H_k P_k^- H_k^T + \Sigma_m \Big)^{-1} \Big( H_k P_k^- \Big)$$

Using the definition of $K_k$ established above:


$$P_k = P_k^- - K_k \Big( H_k P_k^- \Big)$$

Factoring out $P_k^-$ from the right-hand side:


$$P_k = (I - K_k H_k) P_k^-$$

---

#### **Conclusion**

$$x^{+}_k \sim \mathcal{N}(m_k, P_k)$$

where:


$$\begin{aligned}
K_k &= P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1} \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-) \\
P_k &= (I - K_k H_k) P_k^-
\end{aligned}$$



### **Question 5 Solution**

By definition, the conditional expectation $\mathbb{E}[X \mid Y=y]$ corresponds to the mean of the conditional distribution, and the conditional variance $\text{Var}(X \mid Y=y)$ corresponds to the covariance matrix of the conditional distribution.

From the results derived in Question 4, the conditional distribution is known to be:


$$(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathcal{N}(m_k, P_k)$$

Thus, the required conditional expectation and variance are directly given by the updated mean vector $m_k$ and updated covariance matrix $P_k$:

---

#### **1. Conditional Expectation**

$$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}] = m_k$$

$$\mathbf{\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}] = m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-)}$$

where $K_k = P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}$.

---

#### **2. Conditional Variance**

$$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) = P_k$$

$$\mathbf{\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) = (I - K_k H_k) P_k^-}$$

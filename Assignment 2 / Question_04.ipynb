# Question 04

Let $X$ denote the proportion of defective items produced in a large manufacturing batch on a given day.

Since $X$ represents a proportion,

$$ 0\le X\le 1 $$

and $X$ is assumed to follow a Beta distribution with parameters

$$ \alpha=2,\qquad \beta=18 $$

with density

$$ f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\
0, & \text{otherwise}.
\end{cases} $$

   

# 1. Construction of the Probability Space $(\Omega,\mathfrak F,P)$

A suitable probability space is:

$$ \Omega=[0,1] $$

since the defective proportion can take any value between $0$ and $1$.

Let

$$ \mathfrak F=\mathcal B([0,1]) $$

be the Borel $\sigma$-algebra on $[0,1]$.

Define the probability measure $P$ by

$$ P(A)=\int_A f(x)\,dx,
\qquad A\in\mathfrak F $$

where

$$ f(x)=\frac{1}{B(2,18)}x(1-x)^{17}. $$

Hence,

$$ (\Omega,\mathfrak F,P) $$

is the required probability space.



# 2. Definition of the Random Variable

Define

$$ X:\Omega\to\mathbb R $$

by

$$ X(\omega)=\omega,
\qquad \omega\in[0,1]. $$

Thus, $X$ gives the defective proportion in the batch.



# 3. Probability Density Function of $X$

The probability density function is

$$ f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\
0, & \text{otherwise}.
\end{cases} $$

where the Beta function is

$$ B(2,18)=\int_0^1 x^{1}(1-x)^{17}\,dx. $$

---

# 4. Verification that $f$ is a Valid Density Function

To verify that $f$ is a valid probability density function, we check:

### (i) Non-negativity

For $0<x<1$,

$$ x>0,\qquad (1-x)^{17}>0, $$

hence

$$ f(x)\ge0. $$



### (ii) Total Probability Equals 1

We compute

$$ \int_{-\infty}^{\infty} f(x)\,dx = \int_0^1 \frac{1}{B(2,18)}x(1-x)^{17}\,dx. $$

Using the definition of the Beta function,

$$ \int_0^1 x(1-x)^{17}\,dx=B(2,18). $$

Therefore,

$$ \int_0^1 f(x)\,dx = \frac{1}{B(2,18)}B(2,18) = 1. $$

Hence, $f$ is a valid probability density function.


```python
import numpy as np
from scipy.stats import beta
from scipy.special import beta as beta_func # To get the Beta function value
from scipy.integrate import quad

alpha = 2
beta_param = 18

# Calculate the normalizing constant B(alpha, beta)
normalization_constant = beta_func(alpha, beta_param)
print(f"Calculated Beta function B({alpha},{beta_param}) = {normalization_constant:.10f}")

# The PDF function
def f_x(x):
    if 0 < x < 1:
        return (1 / normalization_constant) * (x**(alpha - 1)) * ((1 - x)**(beta_param - 1))
    return 0

integral_value, error = quad(f_x, 0, 1)

print(f"Integral of f(x) from 0 to 1 = {integral_value:.10f}")
print(f"Integration error = {error:.10e}")

if np.isclose(integral_value, 1.0):
    print("The integral is approximately 1.0, so the PDF is valid.")
else:
    print("The integral is not 1.0, there might be an issue.")
```

# 5. Expectation $E[X]$

For a Beta distribution,

$$ E[X]=\frac{\alpha}{\alpha+\beta}. $$

Substituting $\alpha=2$ and $\beta=18$,

$$ E[X]=\frac{2}{2+18} = \frac{2}{20} = 0.10. $$

Therefore,

$$ \boxed{E[X]=0.10} $$



```python
alpha = 2
beta_param = 18

expected_value = alpha / (alpha + beta_param)
print(f"The expectation E[X] = {expected_value:.2f}")

# Verification using scipy.stats
mean_scipy = beta.mean(alpha, beta_param)
print(f"Verification using python: {mean_scipy:.2f}")
```

# 6. Variance $\mathrm{Var}(X)$

For a Beta distribution,

$$ \mathrm{Var}(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}. $$

Substituting $\alpha=2$ and $\beta=18$,

$$ \mathrm{Var}(X) = \frac{(2)(18)}{(20)^2(21)}. $$

$$ \frac{36}{8400} $$

$$ \frac{3}{700} \approx 0.004286. $$

Therefore,

$$ \boxed{\mathrm{Var}(X)=\frac{3}{700}\approx0.004286} $$



```python
alpha = 2
beta_param = 18

variance = (alpha * beta_param) / (((alpha + beta_param)**2) * (alpha + beta_param + 1))
print(f"The variance Var(X) = {variance:.8f}")

# Verification using scipy.stats
var_scipy = beta.var(alpha, beta_param)
print(f"Verification using python: {var_scipy:.8f}")
```

# 7. Compute $P(X\le0.10)$

$$ P(X\le0.10) = \int_0^{0.10} f(x)\,dx $$

Hence,

$$ P(X\le0.10) = \frac{1}{B(2,18)} \int_0^{0.10} x(1-x)^{17}\,dx. $$

Numerically, using the Beta CDF:

$$ P(X\le0.10) \approx 0.57973502 $$

Therefore,

$$ \boxed{
P(X\le0.10) = \frac{1}{B(2,18)} \int_0^{0.10} x(1-x)^{17}\,dx \approx 0.57973502
} $$


```python
alpha = 2
beta_param = 18

# Probability P(X <= 0.10)
prob_le_0_10 = beta.cdf(0.10, alpha, beta_param)
print(f"P(X <= 0.10) = {prob_le_0_10:.8f}")
```

# 8. Compute $P(0.05\le X\le0.15)$

$$ P(0.05\le X\le0.15) = \int_{0.05}^{0.15} f(x)\,dx. $$

Thus,

$$ P(0.05\le X\le0.15) = \frac{1}{B(2,18)} \int_{0.05}^{0.15} x(1-x)^{17}\,dx. $$

Numerically, using the Beta CDF:

$$ P(0.05\le X\le0.15) \approx 0.55621549 $$

Therefore,

$$ \boxed{
P(0.05\le X\le0.15) = \frac{1}{B(2,18)} \int_{0.05}^{0.15} x(1-x)^{17}\,dx \approx 0.55621549
} $$



```python
alpha = 2
beta_param = 18

# Probability P(0.05 <= X <= 0.15)
prob_ge_0_05_le_0_15 = beta.cdf(0.15, alpha, beta_param) - beta.cdf(0.05, alpha, beta_param)
print(f"P(0.05 <= X <= 0.15) = {prob_ge_0_05_le_0_15:.8f}")
```

# 9. Why the Beta Distribution is Reasonable

The Beta distribution is appropriate because:

*   The defective proportion is a continuous quantity restricted to the interval $[0,1]$.
*   The Beta distribution is defined exactly on $[0,1]$.
*   It is flexible and can model many different shapes depending on the parameters $\alpha$ and $\beta$.
*   With $\alpha=2$ and $\beta=18$, the density is concentrated near small values, reflecting that defect rates are usually low in a well-controlled manufacturing process.



# 10. Interpretation of Expectation and Variance

### Expectation

$$ E[X]=0.10 $$

means that the average defective proportion is expected to be $10\%$.

In the long run, about $10\%$ of produced items are expected to be defective.



### Variance

$$ \mathrm{Var}(X)\approx0.004286 $$

measures the variability of the defective proportion from day to day.

The relatively small variance indicates that the production quality is fairly stable, with the defect proportion remaining close to its average value most of the time.

## 1. Bias of the Maximum Likelihood Covariance Estimator

The MLE covariance estimator is

$$\widehat{\Sigma}_{MLE} = \frac{1}{n}\sum_{i=1}^{n}(x_i-\hat{\mu})(x_i-\hat{\mu})^T$$

The unbiased sample covariance matrix is

$$S = \frac{1}{n-1}\sum_{i=1}^{n}(x_i-\hat{\mu})(x_i-\hat{\mu})^T$$

Therefore,

$$\widehat{\Sigma}_{MLE} = \frac{n-1}{n}S$$

Taking expectations,

$$E[\widehat{\Sigma}_{MLE}] = \frac{n-1}{n}E[S]$$

Since

$$E[S]=\Sigma$$

$$\boxed{E[\widehat{\Sigma}_{MLE}] = \frac{n-1}{n}\Sigma}$$

Hence, $\widehat{\Sigma}_{MLE}$ is biased and underestimates the true covariance matrix.

### Bessel's Correction

Multiplying by $\frac{n}{n-1}$,

$$S = \frac{n}{n-1}\widehat{\Sigma}_{MLE}$$

Therefore,

$$E[S] = \frac{n}{n-1}E[\widehat{\Sigma}_{MLE}]$$

$$\frac{n}{n-1}\left(\frac{n-1}{n}\Sigma\right) = \Sigma$$

Hence,

$$\boxed{E[S]=\Sigma}$$

Thus, $S$ is an unbiased estimator of the population covariance matrix.

## 2. Type I Error and Type II Error

### Type I Error ($\alpha$)

Rejecting $H_0$ when $H_0$ is true.

In structural health monitoring:

$$\boxed{\text{Healthy structure declared damaged}}$$

(False alarm)

---

### Type II Error ($\beta$)

Failing to reject $H_0$ when $H_0$ is false.

In structural health monitoring:

$$\boxed{\text{Damaged structure declared healthy}}$$

(Missed detection)

---

### Effect of Choosing ($\alpha=0.0001$)

Reducing $\alpha$ decreases the probability of false alarms.

However,

$$\alpha \downarrow \Rightarrow \beta \uparrow$$

Therefore,

$$\boxed{\text{Statistical Power }(1-\beta)\text{ decreases}}$$

The diagnostic system becomes less sensitive to actual faults.

Since a smaller $\alpha$ requires a larger critical threshold, the healthy-state confidence ellipsoid expands.

$$\boxed{\text{Volume of confidence ellipsoid increases}}$$

As a result, fewer observations are classified as anomalies.

## Slutsky's Theorem

If

$$X_n \xrightarrow{d} X$$

and

$$Y_n \xrightarrow{p} c$$

where $c$ is a constant, then

$$X_n+Y_n \xrightarrow{d} X+c$$

$$X_nY_n \xrightarrow{d} cX$$

$$\frac{X_n}{Y_n} \xrightarrow{d} \frac{X}{c} \qquad (c\neq0)$$

---

## Application of Slutsky's Theorem

From the Multivariate CLT,

$$\sqrt{n}(\hat{\mu}_n-\mu) \xrightarrow{d} N(0,\Sigma)$$

The sample covariance matrix is a consistent estimator:

$$S \xrightarrow{p} \Sigma$$

By Slutsky's Theorem, replacing the unknown $\Sigma$ with $S$ does not change the asymptotic distribution.

Hence,

$$\sqrt{n}(\hat{\mu}_n-\mu) \xrightarrow{d} N(0,S)$$

or equivalently,

$$\boxed{\hat{\mu}_n \sim N\left(\mu, \frac{1}{n}S\right)}$$

Since $\mu$ is unknown in practice, the operational approximation becomes

$$\boxed{\hat{\mu}_n \sim N\left(\hat{\mu}_n, \frac{1}{n}S\right)}$$

which is the practical distribution used for baseline validation and confidence ellipsoid construction.

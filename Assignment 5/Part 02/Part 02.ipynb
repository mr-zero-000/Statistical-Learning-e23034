# 1. Covariance of the Principal Component Coordinates

Given the centered random vector

$$\widetilde{\mathbf{X}}_i=\mathbf{X}_i-\boldsymbol{\mu}$$

with covariance matrix

$$\boldsymbol{\Sigma} = \mathbb{E}\left[\widetilde{\mathbf{X}}_i\widetilde{\mathbf{X}}_i^T\right]$$

and spectral decomposition

$$\boldsymbol{\Sigma} = \mathbf{P}\mathbf{\Lambda}\mathbf{P}^T$$

where

$$\mathbf{P}^T\mathbf{P} = \mathbf{P}\mathbf{P}^T = \mathbf{I},$$

and

$$\mathbf{\Lambda} = \operatorname{diag}(\lambda_1,\lambda_2,\ldots,\lambda_m).$$

The PCA transformation is

$$\mathbf{Z}_i = \mathbf{P}^T\widetilde{\mathbf{X}}_i.$$

---

## Proof of Covariance Transformation

The covariance matrix of $(\mathbf{Z}_i)$ is

$$\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T].$$

Substituting $(\mathbf{Z}_i=\mathbf{P}^T\widetilde{\mathbf{X}}_i)$,

$$\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbb{E} \left[ (\mathbf{P}^T\widetilde{\mathbf{X}}_i) (\mathbf{P}^T\widetilde{\mathbf{X}}_i)^T \right].$$

Using $((AB)^T=B^TA^T)$,

$$\mathbb{E} \left[ \mathbf{P}^T \widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T \mathbf{P} \right].$$

Since $(\mathbf{P})$ is constant,

$$\mathbf{P}^T \mathbb{E} \left[ \widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T \right] \mathbf{P}.$$

Using

$$\mathbb{E} \left[ \widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T \right] = \boldsymbol{\Sigma},$$

$$\mathbf{P}^T\boldsymbol{\Sigma}\mathbf{P}.$$

Substituting

$$\boldsymbol{\Sigma} = \mathbf{P}\mathbf{\Lambda}\mathbf{P}^T,$$

$$\mathbf{P}^T (\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T) \mathbf{P}.$$

Using orthogonality,

$$\mathbf{P}^T\mathbf{P} = \mathbf{I},$$

$$\mathbf{I}\mathbf{\Lambda}\mathbf{I} = \mathbf{\Lambda}.$$

Hence,

$$\boxed{ \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbf{\Lambda} }$$

which is diagonal.

---

## Geometric and Statistical Interpretation

Since

$$\mathbf{\Lambda} = \begin{bmatrix} \lambda_1 & 0 & \cdots & 0\\ 0 & \lambda_2 & \cdots & 0\\ \vdots & \vdots & \ddots & \vdots\\ 0 & 0 & \cdots & \lambda_m \end{bmatrix},$$

the covariance between two different principal components is

$$\operatorname{Cov}(Z_{i,j},Z_{i,k})=0, \qquad j\neq k.$$

Therefore:

### Statistical Meaning

*   Principal components are uncorrelated.
*   Each component contains unique variance information.
*   No linear redundancy exists between different PCs.

### Geometric Meaning

*   The eigenvectors form mutually orthogonal axes.
*   Data are rotated into a coordinate system aligned with directions of maximum variance.
*   Variability along one principal direction does not influence variability along any other direction.

---

# 2. Total Variance Preservation

The total variance in the original space equals

$$\operatorname{tr}(\boldsymbol{\Sigma}).$$

Using

$$\boldsymbol{\Sigma} = \mathbf{P}\mathbf{\Lambda}\mathbf{P}^T,$$

$$\operatorname{tr}(\boldsymbol{\Sigma}) = \operatorname{tr} (\mathbf{P}\mathbf{\Lambda}\mathbf{P}^T).$$

Applying cyclic trace invariance,

$$\operatorname{tr}(ABC) = \operatorname{tr}(BCA),$$

$$\operatorname{tr} (\mathbf{\Lambda}\mathbf{P}^T\mathbf{P}).$$

Since

$$\mathbf{P}^T\mathbf{P} = \mathbf{I},$$

$$\operatorname{tr}(\mathbf{\Lambda}).$$

Since $(\mathbf{\Lambda})$ is diagonal,

$$\operatorname{tr}(\mathbf{\Lambda}) = \sum_{j=1}^{m}\lambda_j.$$

Thus,

$$\boxed{ \operatorname{tr}(\boldsymbol{\Sigma}) = \operatorname{tr}(\mathbf{\Lambda}) = \sum_{j=1}^{m}\lambda_j }$$

showing that PCA preserves total variance.

---

## Cumulative Explained Variance Ratio

For the first $(k)$ principal components,

$$\boxed{ \Phi(k) = \frac{\sum_{j=1}^{k}\lambda_j} {\sum_{j=1}^{m}\lambda_j} }$$

This represents the fraction of total variance captured by the retained subspace.

---

## Residual Unexplained Variance Ratio

The remaining variance is

$$\boxed{ \Psi(k) = \frac{\sum_{j=k+1}^{m}\lambda_j} {\sum_{j=1}^{m}\lambda_j} }$$

Equivalently,

$$\boxed{ \Psi(k)=1-\Phi(k) }$$

which measures variance discarded by dimensionality reduction.

---

# 3. Reconstruction and Residual Analysis

The centered observation is

$$\mathbf{x}_i-\widehat{\boldsymbol{\mu}}_n = \widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}.$$

---

## Reconstructed Vector

Using only the retained principal components,

$$\boxed{ \widehat{\mathbf{x}}_i = \widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}_k\mathbf{z}_{i,k} }$$

---

## Residual Error Vector

The reconstruction error is

$$\mathbf{e}_i = \mathbf{x}_i-\widehat{\mathbf{x}}_i.$$

Substituting,

$$\boxed{ \mathbf{e}_i = \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k} }$$

which lies entirely in the discarded subspace.

---

## Proof of Energy Decomposition

Let

$$\mathbf{y} = \mathbf{x}_i-\widehat{\boldsymbol{\mu}}_n.$$

Then

$$\mathbf{y} = \widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}.$$

Taking squared norm,

$$|\mathbf{y}|^2 = \mathbf{y}^T\mathbf{y}.$$

Substituting,

$$(\widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k})^T (\widehat{\mathbf{P}}_k\mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k}\mathbf{z}_{i,m-k}).$$

Expanding,

$$\mathbf{z}_{i,k}^T \widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \mathbf{z}_{i,m-k}^T \widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$$

$$+ 2\mathbf{z}_{i,k}^T \widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}.$$

Since the eigenvectors are orthonormal,

$$\widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_k = \mathbf{I},$$

$$\widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_{m-k} = \mathbf{I},$$

and

$$\widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_{m-k} = \mathbf{0}.$$

Therefore,

$$|\mathbf{y}|^2 = |\mathbf{z}_{i,k}|^2 + |\mathbf{z}_{i,m-k}|^2.$$

Hence,

$$\boxed{ |\mathbf{x}_i-\widehat{\boldsymbol{\mu}}_n|^2 = |\mathbf{z}_{i,k}|^2 + |\mathbf{z}_{i,m-k}|^2 }$$

which is the Pythagorean decomposition of energy between orthogonal PCA subspaces.

---

# Hotelling's ($T^2$) vs Q (SPE) Statistic

## Hotelling's ($T^2$)

$$T^2 = \mathbf{z}_{i,k}^T \mathbf{\Lambda}_k^{-1} \mathbf{z}_{i,k}$$

Measures variation **within the retained PCA subspace**.

Sensitive to:

*   Large but normal operating variations.
*   Changes that follow historical correlation patterns.
*   Environmental and operational loading changes.

### Example

Extreme wind loads increase structural responses while remaining consistent with learned PCA behavior.

Therefore,

$$\boxed{\text{Environmental load anomaly} \rightarrow T^2}$$

---

## Q Statistic (Squared Prediction Error)

$$Q = |\mathbf{e}_i|^2 = |\mathbf{z}_{i,m-k}|^2.$$

Measures variation in the residual subspace.

Sensitive to:

*   New patterns not represented in baseline data.
*   Structural damage and abnormal dynamics.
*   Changes in load-transfer mechanisms.

### Example

A fatigue crack alters stiffness and redistributes stresses, creating behavior outside the learned PCA model.

Therefore,

$$\boxed{\text{Structural fracture / fatigue crack} \rightarrow Q}$$

---

## Summary

| Statistic | Monitors | Sensitive To |
| :---------------- | :--------------------------- | :------------------------------------------------------------ |
| Hotelling's ($T^2$) | Principal component subspace | Operational/environmental variations, high loading conditions |
| Q (SPE) | Residual subspace | Structural damage, faults, cracks, model mismatch |

Thus, in SHM:

$$\boxed{\text{Extreme wind loading} \rightarrow T^2}$$

$$\boxed{\text{Fatigue crack / structural fracture} \rightarrow Q}$$

because operational changes remain within the learned PCA structure, whereas damage generates new behavior outside the modeled principal subspace.

```python
import numpy as np
import pandas as pd

class PCAHealthMonitoring:

    def evaluate_subspace_statistics(self, X):
        """
        Parameters
        ----------
        X : ndarray of shape (n_samples, n_features)
            Baseline data matrix

        Returns
        -------
        results : pandas.DataFrame
            Contains:
                k               : retained PCs
                mean_T2         : empirical mean Hotelling T²
                mean_Q          : empirical mean Q statistic
                explained_var   : cumulative explained variance ratio
                residual_var    : residual variance ratio
        """

        # --------------------------------------------------
        # 1. Standardize and center the incoming realizations
        # --------------------------------------------------
        mu = np.mean(X, axis=0)
        sigma = np.std(X, axis=0, ddof=1)

        X_std = (X - mu) / sigma

        # --------------------------------------------------
        # 2. Unbiased sample covariance matrix
        # --------------------------------------------------
        S = np.cov(X_std, rowvar=False, ddof=1)

        # Eigen-decomposition
        eigvals, eigvecs = np.linalg.eigh(S)

        # Sort in descending order
        idx = np.argsort(eigvals)[::-1]
        eigvals = eigvals[idx]
        eigvecs = eigvecs[:, idx]

        # PCA scores
        Z = X_std @ eigvecs

        m = X.shape[1]

        results = []

        # --------------------------------------------------
        # 3. Loop over k = 1, 2, ..., m-1
        # --------------------------------------------------
        for k in range(1, m):

            # Retained eigenvalues
            lambdas_k = eigvals[:k]

            # Retained scores
            Zk = Z[:, :k]

            # Residual scores
            Zr = Z[:, k:]

            # ------------------------------------------
            # Hotelling T²
            # T²_i = Σ(z_ij² / λ_j)
            # ------------------------------------------
            T2 = np.sum((Zk ** 2) / lambdas_k, axis=1)

            # ------------------------------------------
            # Q statistic (SPE)
            # Q_i = Σ residual score²
            # ------------------------------------------
            Q = np.sum(Zr ** 2, axis=1)

            # ------------------------------------------
            # 4. Empirical expectations
            # ------------------------------------------
            mean_T2 = np.mean(T2)
            mean_Q = np.mean(Q)

            # Explained / residual variance ratios
            explained_var = np.sum(eigvals[:k]) / np.sum(eigvals)

            residual_var = np.sum(eigvals[k:]) / np.sum(eigvals)

            results.append({
                "k": k,
                "mean_T2": mean_T2,
                "mean_Q": mean_Q,
                "explained_var": explained_var,
                "residual_var": residual_var
            })

        # --------------------------------------------------
        # 5. Return metrics
        # --------------------------------------------------
        return pd.DataFrame(results)
```

```python
# n = 3000 snapshots
# m = 4 sensors

np.random.seed(42)

X = np.random.randn(3000, 4)

monitor = PCAHealthMonitoring()

results = monitor.evaluate_subspace_statistics(X)

print(results)
```

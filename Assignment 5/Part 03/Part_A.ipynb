# 1. Fundamental Equation of Factor Analysis

The Factor Analysis (FA) model assumes that each standardized observation vector

$$ \mathbf{Z}_i \in \mathbb{R}^m $$

can be represented as

$$ \mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i $$

where:

* $\boldsymbol{\Lambda}\in\mathbb{R}^{m\times k}$ is the factor loading matrix.
* $\mathbf{F}_i\in\mathbb{R}^{k}$ is the common factor vector.
* $\boldsymbol{\epsilon}_i\in\mathbb{R}^{m}$ is the unique error vector.

The assumptions are

$$ E[\mathbf{F}_i]=\mathbf{0}, $$

$$ E[\mathbf{F}_i\mathbf{F}_i^T]=\mathbf{I}, $$

$$ E[\boldsymbol{\epsilon}_i]=\mathbf{0}, $$

$$ E[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi} \operatorname{diag} (\varphi_1^2,\varphi_2^2,\ldots,\varphi_m^2), $$

and

$$ E[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = \mathbf{0}. $$

---

## Proof of the Fundamental Equation

The population correlation matrix is

$$ \mathbf{R} = E[\mathbf{Z}_i\mathbf{Z}_i^T]. $$

Substituting the FA model,

$$ \mathbf{R} = E \left[ (\boldsymbol{\Lambda}\mathbf{F}_i+\boldsymbol{\epsilon}_i) (\boldsymbol{\Lambda}\mathbf{F}_i+\boldsymbol{\epsilon}_i)^T \right]. $$

Expanding,

$$ E \left[ \boldsymbol{\Lambda}\mathbf{F}_i\mathbf{F}_i^T\boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}\mathbf{F}_i\boldsymbol{\epsilon}_i^T + \boldsymbol{\epsilon}_i\mathbf{F}_i^T\boldsymbol{\Lambda}^T + \boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T \right]. $$

Applying expectation term-by-term,

$$ \boldsymbol{\Lambda} E[\mathbf{F}_i\mathbf{F}_i^T] \boldsymbol{\Lambda}^T + \boldsymbol{\Lambda} E[\mathbf{F}_i\boldsymbol{\epsilon}_i^T] + E[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] \boldsymbol{\Lambda}^T + E[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]. $$

Using the assumptions,

$$ E[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}, $$

$$ E[\mathbf{F}_i\boldsymbol{\epsilon}_i^T] = \mathbf{0}, $$

$$ E[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = \mathbf{0}, $$

$$ E[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi}, $$

we obtain

$$ \mathbf{R} = \boldsymbol{\Lambda} \mathbf{I} \boldsymbol{\Lambda}^T + \boldsymbol{\Psi}. $$

Hence,

$$ \boxed{ \mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi} } $$

This is known as the **Fundamental Equation of Factor Analysis**.

---

# Communality and Uniqueness

Consider the diagonal entries of

$$ \mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}. $$

For standardized variables,

$$ R_{jj}=1. $$

The $j$-th diagonal element becomes

$$ 1 = \sum_{l=1}^{k}\lambda_{jl}^{2} + \varphi_j^{2}. $$

---

## Communality

The communality of sensor $j$ is

$$ \boxed{ h_j^2 = \sum_{l=1}^{k} \lambda_{jl}^{2} } $$

or

$$ \boxed{ h_j^2 = 1-\varphi_j^2 } $$

for standardized data.

### Interpretation

Communality measures the proportion of sensor variance explained by the common latent factors.

Large $h_j^2$:

* Sensor behavior is strongly driven by global structural dynamics.
* The sensor shares information with the rest of the monitoring network.

Small $h_j^2$:

* Sensor behaves independently.
* The measurement contains little common structural information.

---

## Uniqueness

The uniqueness is

$$ \boxed{ \varphi_j^2 = 1-h_j^2 } $$

and represents variance not explained by common factors.

### Interpretation

Uniqueness contains:

* Measurement noise
* Sensor-specific effects
* Local disturbances
* Sensor faults

Large uniqueness indicates that much of the sensor's variation is isolated from the common structural behavior.

---

## Asset Diagnostics Interpretation

### High Communality

$$ h_j^2 \approx 1 $$

means:

* Sensor tracks global structural response.
* Changes are likely associated with genuine structural behavior.

Example:

* Wind loading affecting the entire bridge.
* Temperature-induced expansion affecting all sensors.

---

### High Uniqueness

$$ \varphi_j^2 \approx 1 $$

means:

* Sensor contains mostly local noise.
* Possible instrumentation issue.

Example:

* Sensor drift.
* Wiring fault.
* Calibration degradation.

Thus communality helps identify **structurally meaningful sensors**, while uniqueness helps identify **sensor-specific abnormalities**.

# 2. Varimax Rotation and Interpretability

---

## Raw Principal Component Loadings

In PCA, loading vectors are chosen to maximize variance.

The first component solves

$$ \max \operatorname{Var}(PC_1), $$

the second solves

$$ \max \operatorname{Var}(PC_2) $$

subject to orthogonality, and so on.

Characteristics:

* Components are ordered by explained variance.
* Loadings are often distributed across many variables.
* Interpretation may be difficult.

Example:

| Sensor | PC1  |
| ------ | ---- |
| S1     | 0.52 |
| S2     | 0.48 |
| S3     | 0.51 |
| S4     | 0.49 |

This captures variance efficiently but does not clearly identify physical mechanisms.

---

## Varimax-Rotated Factor Loadings

Varimax seeks a simpler structure by maximizing loading sparsity.

Typical result:

| Sensor | Factor 1 | Factor 2 |
| ------ | -------- | -------- |
| S1     | 0.90     | 0.05     |
| S2     | 0.88     | 0.08     |
| S3     | 0.06     | 0.91     |
| S4     | 0.04     | 0.89     |

Now each sensor loads strongly on only one factor.

Advantages:

* Easier physical interpretation.
* Clear grouping of sensors.
* Better identification of structural subsystems.

---

# Why Rotation Does Not Change Communality

Suppose an orthogonal rotation matrix

$$ \mathbf{T} $$

satisfies

$$ \mathbf{T}^T\mathbf{T} = \mathbf{I}. $$

The rotated loading matrix is

$$ \boldsymbol{\Lambda}^{*} = \boldsymbol{\Lambda}\mathbf{T}. $$

The covariance reproduced by the rotated model becomes

$$ \boldsymbol{\Lambda}^{*} (\boldsymbol{\Lambda}^{*})^T. $$

Substituting,

$$ (\boldsymbol{\Lambda}\mathbf{T}) (\boldsymbol{\Lambda}\mathbf{T})^T. $$

Using transpose properties,

$$ \boldsymbol{\Lambda} \mathbf{T} \mathbf{T}^T \boldsymbol{\Lambda}^T. $$

Since

$$ \mathbf{T}\mathbf{T}^T = \mathbf{I}, $$

$$ \boldsymbol{\Lambda} \boldsymbol{\Lambda}^T. $$

Therefore,

$$ \boxed{ \boldsymbol{\Lambda}^{*} (\boldsymbol{\Lambda}^{*})^T = \boldsymbol{\Lambda} \boldsymbol{\Lambda}^T } $$

and consequently

$$ \boxed{ \mathbf{R} \approx \boldsymbol{\Lambda}^{*} (\boldsymbol{\Lambda}^{*})^T + \boldsymbol{\Psi} } $$

remains unchanged.

---

# Why Communalities Remain Unchanged

Communality is

$$ h_j^2 = \sum_{l=1}^{k} \lambda_{jl}^{2}. $$

After rotation,

$$ h_j^{*2} = \sum_{l=1}^{k} (\lambda_{jl}^{*})^{2}. $$

Since orthogonal rotation preserves vector length,

$$ |\boldsymbol{\lambda}_j|^2 = |\boldsymbol{\lambda}_j^{*}|^2, $$

where $\boldsymbol{\lambda}_j$ is the $j$-th row of the loading matrix.

Hence

$$ \boxed{ h_j^{*2} = h_j^2 } $$

for every sensor.

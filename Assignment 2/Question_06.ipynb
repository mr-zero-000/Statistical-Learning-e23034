# Question 06

Let $X$ denote the number of days required to complete the software integration task.

Given:

$$a=8,\qquad b=20,\qquad c=11$$

where

* $a$ = optimistic completion time,
* $b$ = pessimistic completion time,
* $c$ = most likely completion time.

The PERT distribution is a scaled Beta distribution on $([a,b])$.



# 1. Compute the Shape Parameters ($\alpha$) and ($\beta$)

Using

$$\alpha = 1+4\left(\frac{c-a}{b-a}\right)$$

and

$$\beta = 1+4\left(\frac{b-c}{b-a}\right),$$

substitute the given values.



### Compute ($\alpha$)

$$\alpha = 1+4\left(\frac{11-8}{20-8}\right)$$

$$\qquad\qquad\qquad = 1+4\left(\frac{3}{12}\right)$$

$$\qquad\qquad\qquad = 1+1 =2.$$

Thus,

<u>$\alpha=2$</u>



### Compute ($\beta$)

$$\beta = 1+4\left(\frac{20-11}{20-8}\right)$$

$$\qquad\qquad\qquad = 1+4\left(\frac{9}{12}\right)$$

$$\qquad\qquad\qquad = 1+3 =4.$$

Thus,

<u>$\beta=4$</u>

# 2. Support of ($X$)

The PERT distribution is defined on the interval

$$[a,b].$$

Hence,

<u>$8\le X\le20$</u>

# 3. Probability Density Function ($f_X(x)$)

The density function is

$$f(x) = \frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}},\qquad a\le x\le b.$$

Substituting

$$a=8,\quad b=20,\quad \alpha=2,\quad \beta=4,$$

gives

$$f_X(x) = \frac{(x-8)^{1}(20-x)^{3}}{B(2,4)(12)^5},\qquad 8\le x\le20.$$

and

$$f_X(x)=0\qquad \text{otherwise}.$$

Therefore,

<u>$f_X(x) = \frac{(x-8)(20-x)^3}{B(2,4)12^5},\qquad 8\le x\le20$</u>

# 4. Probability Space ($( \Omega,\mathfrak F,P)$)

Define

$$\Omega=[8,20]$$

since the completion time lies between (8) and (20) days.

Let

$$\mathfrak F=\mathcal B([8,20])$$

be the Borel ($\sigma$)-algebra on $([8,20])$.

Define the probability measure

$$P(A)=\int_A f_X(x)dx,\qquad A\in\mathfrak F.$$

Hence, the probability space is

<u>$(\Omega,\mathfrak F,P) = ([8,20],\mathcal B([8,20]),P)$</u>

# 5. Definition of the Random Variable

Define

$$X:\Omega\to\mathbb R$$

by

$$X(\omega)=\omega,\qquad \omega\in[8,20].$$

Thus, ($X$) represents the task completion time.

# 6.

$E[X]=\frac{a+4c+b}{6}$

For the PERT distribution, the expected value is defined by the weighted average

$$E[X] = \frac{a+4c+b}{6}.$$

Substituting the given values:

$$E[X] = \frac{8+4(11)+20}{6}$$

$$\qquad\qquad\qquad = \frac{8+44+20}{6}$$

$$\qquad\qquad\qquad = \frac{72}{6} =12.$$

Therefore,

<u>$E[X]=12\text{ days}$</u>

### Proof for $E[X]=\frac{a+4c+b}{6}$

The PERT distribution is a re-scaled Beta distribution. Let $Y \sim \text{Beta}(\alpha, \beta)$ be a standard Beta random variable defined on $[0,1]$.

The PERT random variable $X$ on $[a,b]$ can be related to $Y$ by the transformation:

$$X = a + (b-a)Y.$$

The expected value of $X$ is then:

$$E[X] = E[a + (b-a)Y] = a + (b-a)E[Y].$$

For a Beta distribution $Y \sim \text{Beta}(\alpha, \beta)$, the expected value is:

$$E[Y] = \frac{\alpha}{\alpha+\beta}.$$

Substituting the expressions for $\alpha$ and $\beta$ from the PERT parameters:

$$\alpha = 1+4\left(\frac{c-a}{b-a}\right) = \frac{b-a+4(c-a)}{b-a}$$

$$\beta = 1+4\left(\frac{b-c}{b-a}\right) = \frac{b-a+4(b-c)}{b-a}$$

Now, let's compute $\alpha+\beta$:

$$\alpha+\beta = \frac{b-a+4(c-a) + b-a+4(b-c)}{b-a} = \frac{2(b-a) + 4(c-a+b-c)}{b-a} = \frac{2(b-a) + 4(b-a)}{b-a} = \frac{6(b-a)}{b-a} = 6.$$

So, $E[Y] = \frac{\alpha}{6}$.

Substituting this back into the expression for $E[X]$:

$$E[X] = a + (b-a)\frac{\alpha}{6} = a + (b-a)\frac{1+4\left(\frac{c-a}{b-a}\right)}{6}.$$

$$E[X] = a + \frac{(b-a)}{6} \left(1 + \frac{4(c-a)}{b-a}\right).$$

Distribute $(b-a)/6$:

$$E[X] = a + \frac{b-a}{6} + \frac{4(c-a)}{6}.$$

To combine terms, find a common denominator:

$$E[X] = \frac{6a}{6} + \frac{b-a}{6} + \frac{4c-4a}{6}.$$

$$E[X] = \frac{6a + b - a + 4c - 4a}{6}.$$

Combine like terms ($a$):

$$E[X] = \frac{(6a - a - 4a) + b + 4c}{6} = \frac{a + b + 4c}{6}.$$

Thus, we have shown that:

$$E[X] = \frac{a+4c+b}{6}.$$

# 7. Compute the Variance

Using

$$\mathrm{Var}(X) = \frac{(E[X]-a)(b-E[X])}{7},$$

substitute the values:

$$\mathrm{Var}(X) = \frac{(12-8)(20-12)}{7}$$

$$\qquad\qquad\qquad = \frac{(4)(8)}{7}$$

$$\qquad\qquad\qquad = \frac{32}{7}.$$

Thus,

<u>$\mathrm{Var}(X)=\frac{32}{7}\approx4.57$</u>

# 8. Compute the Standard Deviation

The standard deviation is

$$\sigma=\sqrt{\mathrm{Var}(X)}.$$

Hence,

$$\sigma = \sqrt{\frac{32}{7}}$$

$$\qquad\qquad\qquad \approx2.14.$$

Therefore,

<u>$\sigma\approx2.14\text{ days}$</u>

# 9. Why PERT May Be More Appropriate than a Triangular Distribution

The PERT distribution is often preferred because:

*   it gives more weight to the most likely value ($c$),
*   it produces a smoother density curve,
*   it reduces the effect of extreme values ($a$) and ($b$),
*   it is based on the Beta distribution, which is flexible and realistic for project durations.

In contrast, the Triangular distribution uses straight-line segments and may overemphasize boundary values.

Thus, the PERT model generally provides more stable and realistic project duration estimates.

# 10. Interpretation of ($a$), ($b$), ($c$), ($E[X]$), and ($\sigma$)

### Optimistic Time ($a=8$)

The shortest possible completion time if everything proceeds very efficiently.

---

### Pessimistic Time ($b=20$)

The maximum expected completion time under unfavorable conditions.

---

### Most Likely Time ($c=11$)

The duration considered most realistic by experts.

---

### Expected Time ($E[X]=12$)

The average expected completion time for planning purposes.

Project managers may use this value when preparing schedules and resource allocations.

---

### Standard Deviation ($\sigma\approx2.14$)

Measures the uncertainty or variability in completion time.

A larger standard deviation indicates greater project risk and schedule uncertainty.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang, beta

# Parameters from the problem context
a = 8
b = 20
c_mode = 11

# --- Triangular Distribution Calculations ---
# Scipy's triang.pdf expects 'c' as a normalized value between 0 and 1
c_scipy_triang = (c_mode - a) / (b - a)
loc_triang = a
scale_triang = b - a

# --- PERT Distribution Calculations ---
# Alpha and Beta parameters for PERT distribution (Beta distribution scaled to [a,b])
alpha_pert = 1 + 4 * (c_mode - a) / (b - a)
beta_pert = 1 + 4 * (b - c_mode) / (b - a)

# Generate x-values (time) for plotting
x = np.linspace(a - (b - a) * 0.1, b + (b - a) * 0.1, 500) # Extend range slightly for better visualization

# Calculate PDF values for Triangular distribution
y_triang = triang.pdf(x, c=c_scipy_triang, loc=loc_triang, scale=scale_triang)

# Calculate PDF values for PERT distribution
# The beta.pdf function also takes loc and scale to map to an arbitrary interval [loc, loc + scale]
y_pert = beta.pdf(x, alpha_pert, beta_pert, loc=a, scale=b - a)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot Triangular Distribution
plt.plot(x, y_triang, label='Triangular Distribution', linestyle='--', color='blue')

# Plot PERT Distribution
plt.plot(x, y_pert, label='PERT Distribution', linestyle='-', color='red')

# Add plot details
plt.title(f'Comparison: Triangular vs PERT (a={a}, b={b}, c={c_mode})')
plt.xlabel('Time (days)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.axvline(x=a, color='gray', linestyle=':', label='Optimistic (a)')
plt.axvline(x=b, color='gray', linestyle=':', label='Pessimistic (b)')
plt.axvline(x=c_mode, color='gray', linestyle=':', label='Most Likely (c)')
plt.show()
```

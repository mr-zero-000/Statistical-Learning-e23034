## Question 03

Let $X$ denote the number of emergency transformer failures reported in a single day.

The possible values of $X$ are

$$ \{0,1,2,3\} $$

with probabilities

$$ P(X=0)=0.50,\quad P(X=1)=0.30,\quad P(X=2)=0.15,\quad P(X=3)=0.05 $$

# 1. Construction of the Probability Space $(\Omega,\mathcal{F},P)$

Define the sample space as

$$ \Omega=\{0,1,2,3\} $$

where each outcome represents the number of transformer failures occurring in one day.

Since the sample space is finite, take the sigma-algebra as the power set of $\Omega$:

$$ \mathcal{F}=\mathcal{P}(\Omega) $$

Define the probability measure $P$ by

$$ P(\{0\})=0.50 $$

$$ P(\{1\})=0.30 $$

$$ P(\{2\})=0.15 $$

$$ P(\{3\})=0.05 $$

Check that total probability is 1:

$$ 0.50+0.30+0.15+0.05=1 $$

Hence, $(\Omega,\mathcal{F},P)$ is a valid probability space.

# 2. Definition of the Random Variable

Define the random variable

$$ X:\Omega\to\mathbb{R} $$

by

$$ X(\omega)=\omega $$

for every $\omega\in\Omega$.

Thus, $X$ gives the number of transformer failures in a day.

# 3. Probability Mass Function (PMF)

The probability mass function of $X$ is

$$ p_X(x)=P(X=x) $$

Therefore,

$$ p_X(x)=\begin{cases}
0.50, & x=0 \\
0.30, & x=1 \\
0.15, & x=2 \\
0.05, & x=3 \\
0, & \text{otherwise}
\end{cases} $$

# 4. Expectation ($E[X]$)

The expectation of a discrete random variable is

$$ E[X]=\sum xP(X=x) $$

Hence,

$$ E[X]
=0(0.50)+1(0.30)+2(0.15)+3(0.05)
$$

$$
=0+0.30+0.30+0.15
$$

$$
=0.75
$$

Therefore,

$$ \boxed{E[X]=0.75} $$

# 5. Variance ($\text{Var}(X)$)

First compute $E[X^2]$:

$$ E[X^2]=\sum x^2P(X=x) $$

$$
=0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05)
$$

$$
=0+0.30+0.60+0.45
$$

$$
=1.35
$$

Now use

$$ \text{Var}(X)=E[X^2]-(E[X])^2 $$

$$ \text{Var}(X)=1.35-(0.75)^2 $$

$$
=1.35-0.5625
$$

$$
=0.7875
$$

Therefore,

$$ \boxed{\text{Var}(X)=0.7875} $$

# 6. Find $P(X\ge1)$

$$ P(X\ge1)=P(X=1)+P(X=2)+P(X=3) $$

$$
=0.30+0.15+0.05
$$

$$
=0.50
$$

Thus,

$$ \boxed{P(X\ge1)=0.50} $$

# 7. Find $P(X\ge2)$

$$ P(X\ge2)=P(X=2)+P(X=3) $$

$$
=0.15+0.05
$$

$$
=0.20
$$

Therefore,

$$ \boxed{P(X\ge2)=0.20} $$

## Computations: Expectation, Variance, and Probabilities

```python
import numpy as np

# Define the possible values of X and their probabilities
k_values = np.array([0, 1, 2, 3])
probabilities = np.array([0.50, 0.30, 0.15, 0.05])

# --- Compute Expectation E[X] ---
# E[X] = Σ [k * P(X=k)]
expectation_X = np.sum(k_values * probabilities)
print(f"Expectation E[X]: {expectation_X:.4f}")

# --- Compute E[X^2] for variance calculation ---
# E[X^2] = Σ [k^2 * P(X=k)]
expectation_X_squared = np.sum((k_values**2) * probabilities)
print(f"Expectation E[X^2]: {expectation_X_squared:.4f}")

# --- Compute Variance Var(X) ---
# Var(X) = E[X^2] - (E[X])^2
variance_X = expectation_X_squared - (expectation_X**2)
print(f"Variance Var(X): {variance_X:.4f}")

# --- Find P(X≥1) ---
# P(X≥1) = P(X=1) + P(X=2) + P(X=3) = 1 - P(X=0)
p_X_ge_1 = 1 - probabilities[0] # or np.sum(probabilities[1:])
print(f"P(X≥1): {p_X_ge_1:.4f}")

# --- Find P(X≥2) ---
# P(X≥2) = P(X=2) + P(X=3)
p_X_ge_2 = np.sum(probabilities[2:])
print(f"P(X≥2): {p_X_ge_2:.4f}")
```

# 8. Practical Meaning of Expectation and Variance

### **Expectation**

The expectation

$$ E[X]=0.75 $$

represents the average number of emergency transformer failures expected per day in the district over a long period of time.

Thus, on average, the district experiences about $0.75$ transformer failures per day.


### **Variance**

The variance

$$ Var(X)=0.7875 $$

measures the variability or spread in the daily number of transformer failures.

A larger variance indicates that the number of failures can fluctuate significantly from day to day, while a smaller variance would indicate more consistent daily behavior.

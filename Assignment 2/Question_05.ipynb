# Expected Project Completion Time

## Solution

Let $X$ denote the number of hours required to complete the generator installation.

We are given that $X$ follows a Triangular distribution with parameters

$$a=4,\qquad b=10,\qquad c=6$$

where:

*   $a$ = minimum possible time,
*   $b$ = maximum possible time,
*   $c$ = most likely time (mode).

```python
import sympy
from sympy import integrate, Symbol

# Define parameters
a, b, c = 4, 10, 6
x = Symbol('x')

# Define the piecewise functions for symbolic integration
f_x_part1_sym = 2 * (x - a) / ((b - a) * (c - a))
f_x_part2_sym = 2 * (b - x) / ((b - a) * (b - c))

print(f"Python: Parameters set: a={a}, b={b}, c={c}")
```

### 1. Support of $X$

The support of a Triangular distribution is

$$[a,b].$$

Hence,

$$\underline{4\le X\le 10}$$

```python
# Python Calculation for Support
print(f"Python: The support of X is [{a}, {b}]")
```

### 2. Probability Density Function ($f_X(x)$)

For a Triangular distribution,

$$f_X(x)=
\begin{cases}
\dfrac{2(x-a)}{(b-a)(c-a)}, & a\le x\le c,\\
\dfrac{2(b-x)}{(b-a)(b-c)}, & c<x\le b,\\
0, & \text{otherwise}.
\end{cases}$$

Substituting $a=4$, $b=10$, and $c=6$:

$$f_X(x)=
\begin{cases}
\dfrac{2(x-4)}{(10-4)(6-4)}, & 4\le x\le 6,\\
\dfrac{2(10-x)}{(10-4)(10-6)}, & 6<x\le10,\\
0, & \text{otherwise}.
\end{cases}$$

Simplifying,

$$f_X(x)=
\begin{cases}
\dfrac{x-4}{6}, & 4\le x\le 6,\\
\dfrac{10-x}{12}, & 6<x\le10,\\
0, & \text{otherwise}.
\end{cases}$$

```python
# Python function for numerical PDF evaluation
def f_X_numeric(val, a, b, c):
    if a <= val <= c:
        return 2 * (val - a) / ((b - a) * (c - a))
    elif c < val <= b:
        return 2 * (b - val) / ((b - a) * (b - c))
    else:
        return 0
print("Python: Defined a numerical probability density function f_X_numeric(x).")
print(f"Python: Example f_X(5) = {f_X_numeric(5, a, b, c):.4f}")
```

### 3. Verification that $f_X$ Integrates to 1

We verify

$$\int_{-\infty}^{\infty} f_X(x)\,dx=1.$$

Thus,

$$\int_4^6 \frac{x-4}{6}\,dx
+
\int_6^{10}\frac{10-x}{12}\,dx.$$

First integral:

$$\int_4^6 \frac{x-4}{6}\,dx$$

$$\frac{1}{6}\int_4^6 (x-4)\,dx$$

$$\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_4^6$$

$$\frac{1}{6}\cdot \frac{4}{2} = \frac{1}{3}.$$

Second integral:

$$\int_6^{10}\frac{10-x}{12}\,dx$$

$$\frac{1}{12}\int_6^{10}(10-x)\,dx$$

$$\frac{1}{12}\left[10x-\frac{x^2}{2}\right]_6^{10}$$

$$\frac{1}{12}(50-42) = \frac{8}{12} = \frac{2}{3}.$$

Hence,

$$\frac{1}{3}+\frac{2}{3}=1.$$

Therefore,

$$\underline{\int_{-\infty}^{\infty} f_X(x)\,dx=1}$$

so $f_X$ is a valid density function.

```python
# Python Calculation for integral verification using sympy
integral1_py = integrate(f_x_part1_sym, (x, a, c))
integral2_py = integrate(f_x_part2_sym, (x, c, b))

total_integral_py = integral1_py + integral2_py

print(f"Python: First integral (from {a} to {c}): {integral1_py} (approx {float(integral1_py):.4f})")
print(f"Python: Second integral (from {c} to {b}): {integral2_py} (approx {float(integral2_py):.4f})")
print(f"Python: Total integral: {total_integral_py}. This verifies it integrates to 1.")
```

### 4. Expectation ($E[X]$)

For a Triangular distribution,

$$E[X]=\frac{a+b+c}{3}.$$

Substituting values:

$$E[X]=\frac{4+10+6}{3}$$

$$=\frac{20}{3}$$

$$\approx 6.67.$$

Therefore,

$$\underline{E[X]=\frac{20}{3}\approx6.67\text{ hours}}$$

```python
# Python Calculation for Expectation
expected_value_py = (a + b + c) / 3
print(f"Python: E[X] = ({a} + {b} + {c}) / 3 = {expected_value_py:.4f} hours")
```

### 5. Variance ($\mathrm{Var}(X)$)

For a Triangular distribution,

$$\mathrm{Var}(X)
=
\frac{a^2+b^2+c^2-ab-ac-bc}{18}.$$

Substituting values:

$$\mathrm{Var}(X)
=
\frac{4^2+10^2+6^2-(4)(10)-(4)(6)-(10)(6)}{18}.$$

$$\frac{16+100+36-40-24-60}{18}$$

$$\frac{28}{18} = \frac{14}{9}.$$

Thus,

$$\underline{\mathrm{Var}(X) = \frac{14}{9} \approx 1.56}$$

```python
# Python Calculation for Variance
variance_py = (a**2 + b**2 + c**2 - a*b - a*c - b*c) / 18
print(f"Python: Var(X) = {variance_py:.4f}")
```

### 6. Interpretation of $a$, $b$, and $c$

*   $a=4$: the shortest possible installation time.
*   $b=10$: the longest possible installation time.
*   $c=6$: the most likely installation time.

Thus, the project is expected to most commonly finish around $6$ hours, but could range from $4$ to $10$ hours.

### 7. Why the Triangular Distribution is Reasonable

The Triangular distribution is useful when:

*   only limited information is available,
*   experts can estimate minimum, maximum, and most likely values,
*   exact historical data may not exist.

In project management and construction planning, these three estimates are often available from engineers or contractors, making the Triangular distribution a practical approximation.

### 8. Compute $P(X\le5)$

Since $5\le6$, use the first part of the density:

$$P(X\le5)
=
\int_4^5 \frac{x-4}{6}\,dx.$$

$$\frac{1}{6}\int_4^5 (x-4)\,dx$$

$$\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_4^5$$

$$\frac{1}{6}\cdot \frac{1}{2} = \frac{1}{12}.$$

Therefore,

$$\underline{P(X \le 5) = \frac{1}{12} \approx 0.0833}$$

```python
# Python Calculation for P(X <= 5)
prob_X_le_5_py = integrate(f_x_part1_sym, (x, a, 5))
print(f"Python: P(X <= 5) = {prob_X_le_5_py} (approx {float(prob_X_le_5_py):.4f})")
```

### 9. Compute $P(5\le X\le8)$

Split the integral at the mode ($x=6$):

$$P(5\le X\le8)
=
\int_5^6 \frac{x-4}{6}\,dx
+
\int_6^8 \frac{10-x}{12}\,dx.$$

First integral:

$$\int_5^6 \frac{x-4}{6}\,dx$$

$$\frac{1}{6}\left[\frac{(x-4)^2}{2}\right]_5^6$$

$$\frac{1}{6}\cdot\frac{4-1}{2} = \frac{1}{4}.$$

Second integral:

$$\int_6^8 \frac{10-x}{12}\,dx$$

$$\frac{1}{12}\left[10x-\frac{x^2}{2}\right]_6^8$$

$$\frac{1}{12}(48-42) = \frac{1}{2}.$$

Thus,

$$P(5\le X\le8)
=
\frac{1}{4}+\frac{1}{2} = \frac{3}{4}.$$

Therefore,

$$\underline{P(5 \le X \le 8) = \frac{3}{4} = 0.75}$$

```python
# Python Calculation for P(5 <= X <= 8)
prob_5_le_X_le_6_py = integrate(f_x_part1_sym, (x, 5, c))
prob_6_lt_X_le_8_py = integrate(f_x_part2_sym, (x, c, 8))
total_prob_py = prob_5_le_X_le_6_py + prob_6_lt_X_le_8_py
print(f"Python: P(5 <= X <= 8) = P(5 <= X <= 6) + P(6 < X <= 8) = {prob_5_le_X_le_6_py} + {prob_6_lt_X_le_8_py} = {total_prob_py} (approx {float(total_prob_py):.4f})")
```

### 11. Plot of the Triangular Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values for the plot
x_vals = np.linspace(a - 1, b + 1, 500) # Extend a bit beyond [a, b] to show zero probability

# Calculate y values (PDF) using the defined function
y_vals = [f_X_numeric(val, a, b, c) for val in x_vals]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=f'Triangular PDF (a={a}, b={b}, c={c})', color='blue')

# Highlight the parameters on the x-axis
plt.axvline(a, color='gray', linestyle='--', label=f'Min (a={a})')
plt.axvline(b, color='gray', linestyle='--', label=f'Max (b={b})')
plt.axvline(c, color='red', linestyle='-', label=f'Mode (c={c})')

# Add labels and title
plt.title('Probability Density Function of the Triangular Distribution')
plt.xlabel('Time (x) in hours')
plt.ylabel('Probability Density f(x)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.fill_between(x_vals, 0, y_vals, alpha=0.1, color='blue') # Shade the area under the curve
plt.show()
```

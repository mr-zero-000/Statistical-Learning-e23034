# Question 07

For a PERT distribution, the expected value is

$E[X]=\frac{a+4c+b}{6}.$

We are given:

$a=5,\qquad c=11$

and the condition that the expected value equals the most likely value:

$E[X]=c=11.$

## 1. Determine the Value of (b)

Using the expectation formula,

$11=\frac{5+4(11)+b}{6}.$

Simplify:

$11=\frac{5+44+b}{6}$

$11=\frac{49+b}{6}.$

Multiply both sides by (6):

$66=49+b.$

Hence,

$b=66-49$

$b=17.$

Therefore,

$\underline{b=17\text{ days}}$

## 2. What Happens to the Variance?

For the PERT approximation,

$\mathrm{Var}(X) = \frac{(E[X]-a)(b-E[X])}{7}.$

Since

$E[X]=11,\qquad a=5,\qquad b=17,$

we get

$\mathrm{Var}(X) = \frac{(11-5)(17-11)}{7}.$

$\frac{(6)(6)}{7}$

$\frac{36}{7}.$

Thus,

$\underline{\mathrm{Var}(X)=\frac{36}{7}\approx5.14}$

## Interpretation

Since

$11-5=17-11=6,$

the expected value lies exactly midway between the minimum and maximum times. Therefore, the distribution becomes symmetric around the most likely value.

The variance is still positive, meaning that uncertainty remains in the project duration even though the expected time equals the most likely time.

```python
# Given values
a = 5
c = 11
expected_value = 11 # E[X] = c = 11

# Calculate b
b = 6 * expected_value - a - 4 * c

print(f"The calculated value of b is: {b} days")
```

```python
# Calculate Variance
E_X = expected_value

variance = ((E_X - a) * (b - E_X)) / 7

print(f"The calculated Variance(X) is: {variance}")
print(f"Variance(X) approximately: {variance:.2f}")
```

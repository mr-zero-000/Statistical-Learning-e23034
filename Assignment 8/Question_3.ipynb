# **Question 3**

Let the conditional probability of assigning class $(Y=y)$ given feature vector $(x)$ be $(P(y|x))$.

The objective is to maximize the conditional entropy

$$H(Y|X) = -\sum_x P(x)\sum_y P(y|x)\ln P(y|x)$$

subject to the given feature expectation constraints.

### Step 1: Constraints

The conditional probabilities satisfy

**Normalization**

$$ \sum_{y=0}^{3}P(y|x)=1,\qquad \forall x $$

The observed feature expectations are

$$ E[x_1I(Y=1)]=0.05 $$

$$ E[x_2I(Y=1)]=0.15 $$

$$ E[x_3I(Y=1)]=0.16 $$

which can be written as

$$ \sum_xP(x)x_1P(Y=1|x)=0.05 $$

$$ \sum_xP(x)x_2P(Y=1|x)=0.15 $$

$$ \sum_xP(x)x_3P(Y=1|x)=0.16. $$

---

### Step 2: Form the Lagrangian

Introduce Lagrange multipliers $(\lambda_1,\lambda_2,\lambda_3)$ for the expectation constraints and $(\alpha(x))$ for normalization.

$$ \begin{aligned} L=&-\sum_xP(x)\sum_yP(y|x)\ln P(y|x)\\ &+\lambda_1\left(E[x_1I(Y=1)]-0.05\right)\\ &+\lambda_2\left(E[x_2I(Y=1)]-0.15\right)\\ &+\lambda_3\left(E[x_3I(Y=1)]-0.16\right)\\ &+\sum_x\alpha(x)\left(\sum_yP(y|x)-1\right). \end{aligned} $$

---

### Step 3: Differentiate

Taking the derivative with respect to $(P(y|x))$ and setting it equal to zero gives

$$ P(y|x) \propto \exp\left(\lambda_1x_1I(y=1) +\lambda_2x_2I(y=1) +\lambda_3x_3I(y=1)\right). $$

To ensure the probabilities sum to one, divide by the partition function

$$ Z(x)=\sum_{k=0}^{3} \exp\left(\lambda_1x_1I(k=1) +\lambda_2x_2I(k=1) +\lambda_3x_3I(k=1)\right). $$

Hence,

$$ \boxed{ P(y|x)= \frac{\exp\left(\lambda_1x_1I(y=1)+\lambda_2x_2I(y=1)+\lambda_3x_3I(y=1)\right)} {Z(x)} } $$

---

### Step 4: Logistic Regression Form

The probability of assigning an email to the **Technical** class ($(Y=1))$ is

$$ \boxed{ P(Y=1|x)= \frac{\exp(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)} {Z(x)} } $$

where

* $(\lambda_1<0)$, since financial words rarely indicate Technical support.
* $(\lambda_2>0)$, since account access issues moderately indicate Technical support.
* $(\lambda_3>0)$, with the largest positive value, because bug/malfunction features strongly indicate Technical support.

---

### Conclusion

Therefore, maximizing the conditional entropy under the given feature expectation constraints results in the **multiclass logistic regression (softmax) model**,

$$ \boxed{ P(Y=y|x)= \frac{\exp(\mathbf{w}_y^T\mathbf{x})} {\sum_{k=0}^{3}\exp(\mathbf{w}_k^T\mathbf{x})} } $$

where the weight vector $(\mathbf{w}_y)$ (or equivalently the Lagrange multipliers) is learned so that the expected feature values satisfy the given historical audit constraints. This is the maximum entropy classifier used for multiclass email routing.

# **Question 4**

Let \(S_T\) be the future price of the asset after one month with probability density function \(p(S_T)\), where \(S_T>0\).

The objective is to determine the **Maximum Entropy** distribution subject to the given market constraints.

### Step 1: Define the Entropy

The entropy of a continuous probability distribution is

$$
H(p)=-\int_{0}^{\infty} p(S_T)\ln p(S_T)\,dS_T.
$$

The objective is

$$
\boxed{\text{Maximize } H(p)}
$$

subject to the following constraints.

---

### Step 2: Constraints

**Normalization**

$$
\int_{0}^{\infty} p(S_T)\,dS_T=1
$$

**Expected Future Price**

$$
\int_{0}^{\infty} S_T\,p(S_T)\,dS_T=102
$$

**Expected Squared Price**

$$
\int_{0}^{\infty} S_T^2\,p(S_T)\,dS_T=10600
$$

**Logarithmic Constraint**

$$
\int_{0}^{\infty}\ln(S_T)\,p(S_T)\,dS_T=4.60
$$

---

### Step 3: Form the Lagrangian

Introduce Lagrange multipliers (\(\lambda_0,\lambda_1,\lambda_2,\lambda_3\)).

$$
\begin{aligned}
L=&-\int p(S_T)\ln p(S_T)\,dS_T\\
&+\lambda_0\left(\int p(S_T)\,dS_T-1\right)\\
&+\lambda_1\left(\int S_Tp(S_T)\,dS_T-102\right)\\
&+\lambda_2\left(\int S_T^2p(S_T)\,dS_T-10600\right)\\
&+\lambda_3\left(\int \ln(S_T)p(S_T)\,dS_T-4.60\right).
\end{aligned}
$$

---

### Step 4: Differentiate

Differentiate with respect to (\(p(S_T)\)) and set equal to zero.

$$
\frac{\partial L}{\partial p(S_T)}
=-\ln p(S_T)-1+\lambda_0+\lambda_1S_T+\lambda_2S_T^2+\lambda_3\ln(S_T)=0.
$$

Hence,

$$
\ln p(S_T)
=\lambda_0-1+\lambda_1S_T+\lambda_2S_T^2+\lambda_3\ln(S_T).
$$

Taking the exponential of both sides,

$$
p(S_T)
=A\,S_T^{\lambda_3}
\exp\left(\lambda_1S_T+\lambda_2S_T^2\right),
$$

where

$$
A=e^{\lambda_0-1}.
$$

For the density to remain finite,

$$
\lambda_2<0.
$$

---

### Step 5: Final Maximum Entropy Distribution

The maximum entropy probability density function is

$$
\boxed{
p(S_T)
=\frac{1}{Z}
S_T^{\lambda_3}
\exp\left(\lambda_1S_T+\lambda_2S_T^2\right),
\qquad S_T>0
}
$$

where

$$
Z=\int_{0}^{\infty}
S_T^{\lambda_3}
\exp\left(\lambda_1S_T+\lambda_2S_T^2\right)dS_T
$$

is the normalization constant, and the multipliers (\(\lambda_1,\lambda_2,\lambda_3\)) are determined such that

$$
E[S_T]=102,\qquad
E[S_T^2]=10600,\qquad
E[\ln(S_T)]=4.60.
$$

---

### Conclusion

The Maximum Entropy solution is

$$
\boxed{
p(S_T)=\frac{1}{Z}
S_T^{\lambda_3}
\exp\left(\lambda_1S_T+\lambda_2S_T^2\right),
\qquad S_T>0.
}
$$

This distribution satisfies all the market constraints while making the fewest assumptions about the unknown future asset price, making it suitable for **asset pricing, derivative valuation, and portfolio risk optimization**.

# **Question 01**

# **Statistical Mechanics of the Quantum Harmonic Oscillator**

Let

$$E_j=\hbar\omega\left(j+\frac12\right), \qquad j=0,1,2,\ldots$$

Define

$$\beta=\frac{1}{k_BT}.$$



# **1. Partition Function**

The canonical partition function is

$$Z(T)=\sum_{j=0}^{\infty}e^{-\beta E_j}.$$

Substituting the energy eigenvalues,

$$\begin{aligned}
Z
&=\sum_{j=0}^{\infty}
e^{-\beta\hbar\omega\left(j+\frac12\right)}
\\
&=e^{-\beta\hbar\omega/2}
\sum_{j=0}^{\infty}
\left(e^{-\beta\hbar\omega}\right)^j.
\end{aligned}$$

Since

$$e^{-\beta\hbar\omega}<1,$$

the series is a convergent geometric series,

$$\sum_{j=0}^{\infty}r^j=\frac{1}{1-r},
\qquad |r|<1.$$

Hence,

$$\boxed{
Z(T)=
\frac{e^{-\beta\hbar\omega/2}}
{1-e^{-\beta\hbar\omega}}
}$$

or equivalently,

$$\boxed{
Z(T)=
\frac{e^{-\frac{\hbar\omega}{2k_BT}}}
{1-e^{-\frac{\hbar\omega}{k_BT}}}.
}$$



# **2. Thermodynamic Potentials**

## (a) Helmholtz Free Energy

The Helmholtz free energy is

$$F=-k_BT\ln Z.$$

Using

$$\ln Z
=-\frac{\beta\hbar\omega}{2}
-\ln\left(1-e^{-\beta\hbar\omega}\right),$$

gives

$$\begin{aligned}
F
&=
-k_BT
\left[
-\frac{\beta\hbar\omega}{2}
-\ln\left(1-e^{-\beta\hbar\omega}\right)
\right]
\\
&=
\frac{\hbar\omega}{2}
+k_BT
\ln\left(1-e^{-\beta\hbar\omega}\right).
\end{aligned}$$

Therefore,

$$\boxed{
F(T)=
\frac{\hbar\omega}{2}
+k_BT
\ln\left(1-e^{-\frac{\hbar\omega}{k_BT}}\right)
}$$



## (b) Internal Energy

The internal energy is

$$U=-\frac{\partial\ln Z}{\partial\beta}.$$

Differentiate

$$\ln Z
=-\frac{\beta\hbar\omega}{2}
-\ln\left(1-e^{-\beta\hbar\omega}\right).$$

Hence,

$$\frac{\partial\ln Z}{\partial\beta}
=-\frac{\hbar\omega}{2}
+\frac{\hbar\omega e^{-\beta\hbar\omega}}
{1-e^{-\beta\hbar\omega}}.$$

Therefore,

$$\begin{aligned}
U
&=-\frac{\partial\ln Z}{\partial\beta}
\\
&=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}.
\end{aligned}$$

Thus,

$$\boxed{
U(T)=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\frac{\hbar\omega}{k_BT}}-1}
}$$



# **3. Entropy**

The entropy is

$$S
=-\left(\frac{\partial F}{\partial T}\right).$$

From

$$F=
\frac{\hbar\omega}{2}
+
k_BT
\ln\left(1-e^{-\beta\hbar\omega}\right),$$

differentiate with respect to (T),

$$\boxed{
S
=
k_B
\left[
\frac{\beta\hbar\omega}
{e^{\beta\hbar\omega}-1}
-\ln\left(1-e^{-\beta\hbar\omega}\right)
\right].
}$$

Substituting

$$F=
\frac{\hbar\omega}{2}
+
k_BT
\ln(1-e^{-\beta\hbar\omega})$$

and

$$TS
=\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}
-k_BT
\ln(1-e^{-\beta\hbar\omega}),$$

gives

$$\begin{aligned}
F+TS
&=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}
\\
&=U.
\end{aligned}$$

Hence,

$$\boxed{U=F+TS}$$

is verified.



# **4. Heat Capacity**

The heat capacity at constant volume is

$$C_V
=\left(\frac{\partial U}{\partial T}\right).$$

Using

$$U
=\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1},$$

let

$$x=\frac{\hbar\omega}{k_BT}.$$

Then

$$U
=\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}{e^x-1}.$$

Since

$$\frac{dx}{dT}
=-\frac{x}{T},$$

we obtain

$$\boxed{
C_V
=
k_B
\left(\frac{\hbar\omega}{k_BT}\right)^2
\frac{e^{\frac{\hbar\omega}{k_BT}}}
{\left(e^{\frac{\hbar\omega}{k_BT}}-1\right)^2}
}$$



## Hyperbolic Form

Using

$$\sinh(y)=\frac{e^y-e^{-y}}{2},$$

let

$$y=\frac{\hbar\omega}{2k_BT},
\qquad x=2y.$$

Since

$$e^x-1
=2e^{x/2}\sinh\left(\frac{x}{2}\right),$$

then

$$(e^x-1)^2
=4e^x\sinh^2\left(\frac{x}{2}\right).$$

Therefore,

$$\frac{e^x}{(e^x-1)^2}
=\frac{1}
{4\sinh^2(x/2)}.$$

Substituting into the previous expression,

$$\boxed{
C_V
=
k_B
\left(
\frac{\hbar\omega}{2k_BT}
\right)^2
\frac{1}
{\sinh^2\left(\frac{\hbar\omega}{2k_BT}\right)}
}$$

which is the required compact form.



# **5. Asymptotic Behaviour (Discussion)**

In the high-temperature limit,

$$k_BT\gg\hbar\omega,$$

or

$$x=\frac{\hbar\omega}{2k_BT}\ll1.$$

Using the approximation

$$\sinh x\approx x,$$

the heat capacity becomes

$$\begin{aligned}
C_V
&\approx
k_B
\frac{x^2}{x^2}
\\
&=k_B.
\end{aligned}$$

Hence,

$$\boxed{C_V\rightarrow k_B
\qquad
(T\rightarrow\infty)}$$

This agrees exactly with the **classical equipartition theorem**, which predicts that a one-dimensional harmonic oscillator has two quadratic energy terms (kinetic and potential), each contributing ($\tfrac{1}{2}k_BT$). Therefore,

$$U\approx k_BT,$$

and

$$\boxed{C_V=\frac{dU}{dT}=k_B.}$$

Thus, the quantum result correctly approaches the classical limit at high temperatures.

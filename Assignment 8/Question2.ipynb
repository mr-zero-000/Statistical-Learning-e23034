# Question 2

To find the probability distribution that maximizes entropy, we start with the entropy of a continuous random variable:

$$
H(f)=-\int_{-\infty}^{\infty} f(x)\ln f(x)\,dx
$$

The distribution must satisfy the following constraints:

1.  **Normalization**

$$
\int_{-\infty}^{\infty} f(x)\,dx=1
$$

2.  **Mean**

$$
\int_{-\infty}^{\infty} x f(x)\,dx=\mu
$$

3.  **Variance**

$$
\int_{-\infty}^{\infty} (x-\mu)^2f(x)\,dx=\sigma^2
$$

Using the **method of Lagrange multipliers**, define the function

$$
\begin{aligned}
L=&- \int f(x)\ln f(x)\,dx
+\lambda_0\left(\int f(x)\,dx-1\right)\\
&+\lambda_1\left(\int xf(x)\,dx-\mu\right)
+\lambda_2\left(\int (x-\mu)^2f(x)\,dx-\sigma^2\right).
\end{aligned}
$$

Differentiate with respect to $f(x)$ and set the result equal to zero:

$$
\frac{\partial L}{\partial f(x)}
=-\ln f(x)-1+\lambda_0+\lambda_1x+\lambda_2(x-\mu)^2=0.
$$

Hence,

$$
\ln f(x)=\lambda_0-1+\lambda_1x+\lambda_2(x-\mu)^2.
$$

Taking the exponential of both sides,

$$
f(x)=A\exp\left(\lambda_1x+\lambda_2(x-\mu)^2\right),
$$

where $A=e^{\lambda_0-1}$.

Since the density must be finite as $x\rightarrow\pm\infty$, $\lambda_2<0$. Applying the mean and variance constraints gives

$$
\lambda_1=0,\qquad
\lambda_2=-\frac{1}{2\sigma^2},
$$

and the normalization condition gives

$$
A=\frac{1}{\sqrt{2\pi\sigma^2}}.
$$

Therefore,

$$
\boxed{
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
}
$$

Thus, the probability distribution that **maximizes entropy subject to fixed mean $\mu$ and variance $\sigma^2$** is the **Gaussian (Normal) distribution**:

$$
\boxed{
X\sim\mathcal{N}(\mu,\sigma^2)
}
$$

**Conclusion:** The maximum entropy distribution under the given mean and variance constraints is the normal distribution,

$$
\boxed{
X\sim\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right).
}
$$

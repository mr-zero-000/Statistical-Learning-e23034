Let the sample space be:

$$ S = \{BB, BG, GB, GG\} $$

Assumption: each outcome is equally likely.

So,
$$ P(BB)=P(BG)=P(GB)=P(GG)=\frac{1}{4} $$



### But we are given like :

At least one child is a girl

So we remove (BB):

$$ S' = \{BG, GB, GG\} $$

$$ P(S') = \frac{3}{4} $$


so then
### Required probability:

Both children are girls given at least one is a girl:

$$ P(GG \mid \text{at least one girl}) = \frac{P(GG)}{P(S')} $$

$$ = \frac{\frac{1}{4}}{\frac{3}{4}} = \frac{1}{3} $$

---

### Final Answer:

$$ \boxed{\frac{1}{3}} $$

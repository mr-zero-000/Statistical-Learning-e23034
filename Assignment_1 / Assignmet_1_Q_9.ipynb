# **We are Given**

$$\Omega=\{G_1,G_2,D_1,D_2\}$$

$$P(G_1)=0.50,\; P(G_2)=0.20,\; P(D_1)=0.10,\; P(D_2)=0.20$$

$$0.50+0.20+0.10+0.20=1$$

So (P) is a valid probability measure.


# **1. Probability space**

$$(\Omega,\mathfrak F,P)$$

where

$$\mathfrak F=\mathcal P(\Omega)$$



# **2. $\sigma$-algebra generated by (X)**

$$X(G_1)=X(G_2)=0,\quad X(D_1)=X(D_2)=1$$

So:

$$X^{-1}(0)=\{G_1,G_2\},\quad X^{-1}(1)=\{D_1,D_2\}$$

Hence:

$$\sigma(X)=\{\emptyset,\Omega,\{G_1,G_2\},\{D_1,D_2\}\}$$



# **3. $\sigma$-algebra generated by (Y)**

$$Y(G_1)=1,\; Y(G_2)=2,\; Y(D_1)=3,\; Y(D_2)=4$$

Each outcome is uniquely identified, so:

$$\sigma(Y)=\mathcal P(\Omega)$$



# **4. Show $(\sigma(X)\subseteq\sigma(Y))$**

Since $(\sigma(Y))$ is the power set of $(\Omega)$, it contains all subsets, including:

$$\emptyset,\Omega,\{G_1,G_2\},\{D_1,D_2\}$$

Thus:

$$\sigma(X)\subseteq\sigma(Y)$$

**Interpretation:**
(Y) gives full information about outcome, while (X) only distinguishes good vs defective.



# **5. Marginal distribution of (X)**

$$P(X=0)=P(G_1)+P(G_2)=0.70$$

$$P(X=1)=P(D_1)+P(D_2)=0.30$$



# **6. Marginal distribution of (Y)**

$$P(Y=1)=0.50,\quad P(Y=2)=0.20,\quad P(Y=3)=0.10,\quad P(Y=4)=0.20$$



# **7. Joint probabilities**

$$P(X=0,Y=1)=0.50$$
$$P(X=0,Y=2)=0.20$$
$$P(X=1,Y=3)=0.10$$
$$P(X=1,Y=4)=0.20$$

All others are zero.



# **8. Conditional probabilities**

### (i)

$$P(X=1\mid Y=3)=1$$

### (ii)

$$P(X=1\mid Y=4)=1$$

### (iii)

$$P(Y=3\mid X=1)=\frac{0.10}{0.30}=\frac{1}{3}$$

### (iv)

$$P(Y=4\mid X=1)=\frac{0.20}{0.30}=\frac{2}{3}$$



# **9. Interpretation of conditioning**

*   Conditioning on (Y): gives complete certainty about outcome $\Rightarrow$ no ambiguity.
*   Conditioning on (X): only tells defective/good $\Rightarrow$ machine identity still uncertain.

Thus, (Y) provides stronger information than (X).



# **10. Information in $\sigma$-algebras**

*   ($\sigma(X)$): partitions space into

    *   Good: $(\{G_1,G_2\})$
    *   Defective: $(\{D_1,D_2\})$

*   ($\sigma(Y)$): separates all outcomes individually.

So:

*   ($\sigma(X)$): coarse information (classification only)
*   ($\sigma(Y)$): complete microscopic information

# **12. Shannon Entropy (natural logs)**

## Marginal entropy of (X)

$$H(X)=-(0.7\ln0.7 + 0.3\ln0.3)$$

$$H(X)\approx 0.6109$$


## Marginal entropy of (Y)

$$H(Y)=-(0.5\ln0.5 + 0.2\ln0.2 + 0.1\ln0.1 + 0.2\ln0.2)$$

$$H(Y)\approx 1.2207$$


## Joint entropy

Same distribution over $(\Omega)$:

$$H(X,Y)\approx 1.2207$$



## Conditional entropies

$$H(X|Y)=0$$

$$H(Y|X)=H(X,Y)-H(X)=1.2207-0.6109=0.6098$$



# **Final Summary**

$$H(X)=0.6109$$
$$H(Y)=1.2207$$
$$H(X,Y)=1.2207$$
$$H(X|Y)=0$$
$$H(Y|X)=0.6098$$



# **Final Interpretation**

*   (X) = compressed information (lossy)
*   (Y) = full information (lossless)
*   Therefore:
    $$\sigma(X)\subset \sigma(Y),\quad H(X)<H(Y)$$


````python
# Question 11
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import entropy

# --- Data Definition (from the problem statement) ---
data = {
    'outcome': ['G1', 'G2', 'D1', 'D2'],
    'P(omega)': [0.50, 0.20, 0.10, 0.20],
    'X_value': [0, 0, 1, 1], # X: 0=Good, 1=Defective
    'Y_value': [1, 2, 3, 4]  # Y: 1=G1, 2=G2, 3=D1, 4=D2
}
df = pd.DataFrame(data)

# --- 11. Python Data Visualizations ---
print("--- Python Data Visualizations ---")

# Calculate Marginal Distributions
P_X = df.groupby('X_value')['P(omega)'].sum().reset_index()
P_Y = df.groupby('Y_value')['P(omega)'].sum().reset_index()

# Calculate Joint Distribution (P(X,Y))
# This is directly from the P(omega) where X and Y values align
joint_matrix = df.pivot_table(index='X_value', columns='Y_value', values='P(omega)', fill_value=0)

# --- Plotting ---

# Plot 1: Marginal Distribution of X
plt.figure(figsize=(6, 4))
sns.barplot(x='X_value', y='P(omega)', data=P_X, palette='viridis')
plt.title('Marginal Distribution of X (Quality)')
plt.xlabel('X (0=Good, 1=Defective)')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Good (X=0)', 'Defective (X=1)'])
for index, row in P_X.iterrows():
    plt.text(row.name, row['P(omega)'] + 0.02, f"{row['P(omega)']:.2f}", color='black', ha="center")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# Plot 2: Marginal Distribution of Y
plt.figure(figsize=(8, 4))
sns.barplot(x='Y_value', y='P(omega)', data=P_Y, palette='plasma')
plt.title('Marginal Distribution of Y (Quality & Machine)')
plt.xlabel('Y (1=G1, 2=G2, 3=D1, 4=D2)')
plt.ylabel('Probability')
plt.xticks([0, 1, 2, 3], ['G1 (Y=1)', 'G2 (Y=2)', 'D1 (Y=3)', 'D2 (Y=4)'])
for index, row in P_Y.iterrows():
    plt.text(row.name, row['P(omega)'] + 0.02, f"{row['P(omega)']:.2f}", color='black', ha="center")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

# Plot 3: Joint Distribution P(X,Y) as a Heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(joint_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5, linecolor='black')
plt.title('Joint Probability Distribution P(X,Y)')
plt.xlabel('Y Value')
plt.ylabel('X Value')
plt.tight_layout()
plt.show()

# Plot 4: Conditional Probabilities P(Y|X=1)
# The conditional probabilities P(Y=3|X=1) and P(Y=4|X=1) were 1/3 and 2/3 respectively.
conditional_y_given_x1_probs = [
    df[(df['X_value'] == 1) & (df['Y_value'] == 3)]['P(omega)'].sum() / P_X[P_X['X_value']==1]['P(omega)'].iloc[0],
    df[(df['X_value'] == 1) & (df['Y_value'] == 4)]['P(omega)'].sum() / P_X[P_X['X_value']==1]['P(omega)'].iloc[0]
]
conditional_y_given_x1_labels = ['P(Y=3|X=1)', 'P(Y=4|X=1)']

plt.figure(figsize=(6, 4))
sns.barplot(x=conditional_y_given_x1_labels, y=conditional_y_given_x1_probs, palette='Reds')
plt.title('Conditional Probabilities P(Y|X=1)')
plt.xlabel('Event')
plt.ylabel('Probability')
for i, val in enumerate(conditional_y_given_x1_probs):
    plt.text(i, val + 0.02, f"{val:.2f}", color='black', ha="center")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()


# --- 12. Shannon Entropy Calculations (base 2 for bits) ---
print("\n--- Shannon Entropy Calculations (base 2 bits) ---")

# Marginal Entropy H(X)
H_X = entropy(P_X['P(omega)'], base=2)
print(f"H(X) (Marginal Entropy of X): {H_X:.4f} bits")

# Marginal Entropy H(Y)
H_Y = entropy(P_Y['P(omega)'], base=2)
print(f"H(Y) (Marginal Entropy of Y): {H_Y:.4f} bits")

# Joint Entropy H(X,Y)
# The joint probabilities are the probabilities of the elementary events themselves.
# So H(X,Y) = H(P(omega))
H_XY = entropy(df['P(omega)'], base=2)
print(f"H(X,Y) (Joint Entropy of X and Y): {H_XY:.4f} bits")

# Conditional Entropy H(X|Y)
# H(X|Y) = H(X,Y) - H(Y)
H_X_given_Y = H_XY - H_Y
print(f"H(X|Y) (Conditional Entropy of X given Y): {H_X_given_Y:.4f} bits")

# Conditional Entropy H(Y|X)
# H(Y|X) = H(X,Y) - H(X)
H_Y_given_X = H_XY - H_X
print(f"H(Y|X) (Conditional Entropy of Y given X): {H_Y_given_X:.4f} bits")
````

$$\textbf{Answer for Q3}$$

Given:

$$X \sim \mathcal N(0,1),$$

$$Q \sim \mathcal N(0,4),$$

where $X$ and $Q$ are independent, and

$$Y=X+Q.$$



# 1. Joint Distribution

Since $X$ and $Q$ are Gaussian and independent, the linear combination

$$Y=X+Q$$

is also Gaussian.

Hence,

$$(X,Y)$$

has a joint multivariate normal distribution.



## Distribution of $Y$

Mean:

$$E[Y]=E[X+Q]$$

$$=E[X]+E[Q]$$

$$=0+0=0.$$

Variance:

$$\text{Var}(Y)
=
\text{Var}(X+Q).$$

Since $X$ and $Q$ are independent,

$$\text{Var}(Y)
=
\text{Var}(X)+\text{Var}(Q)$$

$$=1+4=5.$$

Therefore,

$$\boxed{
Y\sim \mathcal N(0,5)
}$$



## Covariance $(\text{Cov}(X,Y))$

$$\text{Cov}(X,Y)
=
\text{Cov}(X,X+Q).$$

Using linearity,

$$\text{Cov}(X,X)+\text{Cov}(X,Q).$$

Since $X$ and $Q$ are independent,

$$\text{Cov}(X,Q)=0.$$

Thus,

$$\text{Cov}(X,Y)=\text{Var}(X)=1.$$

Hence,

$$\boxed{
\text{Cov}(X,Y)=1
}$$



# 2. Information Geometry

A random variable is $\sigma(Y)$-measurable if it can be completely determined from observing $Y$.

However,

$$Y=X+Q,$$

and the engineer only observes the sum.

Suppose the power meter shows

$$Y=5.$$

This could happen because:

* the temperature is high ($X$ large),
* the cutting torque is high ($Q$ large),
* or both contribute together.

For example,

$$5=5+0,$$

or

$$5=2+3,$$

or

$$5=0+5.$$

Thus observing $Y$ does not uniquely determine $X$.

Therefore,

$$\boxed{
X \text{ is not } \sigma(Y)\text{-measurable}
}$$

because the hidden noise $Q$ prevents exact recovery of the temperature.



# 3. Conditional Expectation $(E[X|Y])$

For jointly Gaussian variables,

$$E[X|Y=y]
=
\mu_X
+
\frac{\text{Cov}(X,Y)}
{\text{Var}(Y)}
(y-\mu_Y).$$

Since

$$\mu_X=0,
\qquad
\mu_Y=0,$$

$$\text{Cov}(X,Y)=1,
\qquad
\text{Var}(Y)=5,$$

we obtain

$$E[X|Y=y]
=
\frac15 y.$$

Hence,

$$\boxed{
E[X|Y]
=
\frac15 Y
}$$



## Signal-to-Total-Variance Ratio

The estimator can be written as

$$E[X|Y]
=
\frac{\text{signal variance}}
{\text{total variance}}
Y.$$

Here:

$$\text{signal variance}=1,$$

$$\text{noise variance}=4,$$

$$\text{total variance}=1+4=5.$$

Therefore,

$$E[X|Y]
=
\frac{1}{1+4}Y.$$

Thus only one-fifth of the observed power is attributed to temperature.



# 4. Verification of the Residual

Let

$$\widehat X=E[X|Y]=\frac15Y.$$

Define the residual

$$Z=X-\widehat X.$$

Thus,

$$Z=X-\frac15Y.$$



## Show that $(E[Z]=0)$

$$E[Z]
=
E\left[X-\frac15Y\right].$$

Using linearity,

$$E[X]-\frac15E[Y].$$

Since

$$E[X]=0,
\qquad
E[Y]=0,$$

we obtain

$$\boxed{
E[Z]=0
}$$



## Orthogonality Principle

We verify that

$$\text{Cov}(Z,Y)=0.$$

Compute:

$$\text{Cov}(Z,Y)
=
\text{Cov}\left(X-\frac15Y,Y\right).$$

Using linearity,

$$\text{Cov}(X,Y)
-\frac15\text{Var}(Y).$$

Substituting values:

$$1-\frac15(5)$$

$$=1-1=0.$$

Hence,

$$\boxed{
\text{Cov}(Z,Y)=0
}$$

Therefore the estimation error is orthogonal to the observation.



# 5. Differential Entropy

For a Gaussian random variable,

$$h(X)
=
\frac12\log(2\pi e\sigma^2).$$



## Entropy of $X$

Since

$$X\sim\mathcal N(0,1),$$

we have

$$h(X)
=
\frac12\log(2\pi e).$$

Thus,

$$\boxed{
h(X)=\frac12\log(2\pi e)
}$$



## Conditional Differential Entropy $(h(X|Y))$

The conditional variance is

$$\sigma_{X|Y}^2
=
\sigma_X^2
-
\frac{\text{Cov}(X,Y)^2}
{\text{Var}(Y)}.$$

Substituting values:

$$1-\frac{1^2}{5}$$

$$1-\frac15
=
\frac45.$$

Hence,

$$X|Y
\sim
\mathcal N\left(\frac15Y,\frac45\right).$$

Therefore,

$$h(X|Y)
=
\frac12
\log\left(2\pi e\cdot \frac45\right).$$

Thus,

$$\boxed{
h(X|Y)
=
\frac12\log\left(\frac45 2\pi e\right)
}$$



# 6. Mutual Information

Mutual information is

$$I(X;Y)=h(X)-h(X|Y).$$

Substituting results:

$$I(X;Y)
=
\frac12\log(2\pi e)
-
\frac12\log\left(2\pi e\cdot \frac45\right).$$

Using logarithm rules,

$$I(X;Y)
=
\frac12\log\left(\frac54\right).$$

Hence,

$$\boxed{
I(X;Y)=\frac12\log\left(\frac54\right)
}$$

Numerically,

$$I(X;Y)\approx0.112 \text{ nats}.$$



## Percentage of Uncertainty Removed

The fraction of uncertainty removed is

$$1-\frac{h(X|Y)}{h(X)}.$$

Since

$$h(X|Y)<h(X),$$

observing $Y$ reduces uncertainty about temperature.



## If Noise Variance Becomes 100

Suppose

$$Q\sim\mathcal N(0,100).$$

Then

$$\text{Var}(Y)=1+100=101.$$

The estimator becomes

$$E[X|Y]
=
\frac1{101}Y.$$

The conditional variance becomes

$$\sigma_{X|Y}^2
=
1-\frac1{101}
=
\frac{100}{101}.$$

Thus the uncertainty reduction becomes extremely small.

Therefore the power meter becomes much less informative when measurement noise is very large.



# 7. Diagnostic Interpretation

The manager claims:

$$\text{Best estimate of }X\text{ is }Y.$$

However,

$$Y=X+Q.$$

Since $Q$ contains random noise, $Y$ overreacts to fluctuations caused by mechanical load.

The optimal estimator is actually

$$E[X|Y]=\frac15Y.$$

This estimator shrinks the observation toward zero because only part of the measured power is due to temperature.

For example, if

$$Y=10,$$

the best estimate is

$$E[X|Y=10]=2,$$

not $10$.

This occurs because most variability in $Y$ comes from noise:

$$\text{Noise variance}=4
>
\text{Signal variance}=1.$$

Hence the conditional expectation filters out noise and produces a more reliable estimate.

Therefore,

$$\boxed{
Y \text{ is a noisy and biased estimator of }X,
}$$

while

$$\boxed{
E[X|Y]
}$$

is the minimum mean-square-error estimator.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import norm

# ==========================================================
# PYTHON VISUALIZATION FOR Q3
# Continuous Fault Monitoring of a Spindle Motor
# ==========================================================

# ----------------------------------------------------------
# 1. Problem Parameters
# ----------------------------------------------------------
sigma_x2 = 1          # variance of temperature
sigma_q2 = 4          # variance of mechanical load noise

sigma_y2 = sigma_x2 + sigma_q2

cov_xy = sigma_x2

# MMSE shrinkage coefficient
alpha = cov_xy / sigma_y2

# conditional variance
sigma_cond2 = sigma_x2 - cov_xy**2 / sigma_y2

# ----------------------------------------------------------
# 2. Generate Monte Carlo Data
# ----------------------------------------------------------
np.random.seed(10)

N = 4000

X = np.random.normal(0, np.sqrt(sigma_x2), N)
Q = np.random.normal(0, np.sqrt(sigma_q2), N)

Y = X + Q

# MMSE estimate
X_hat = alpha * Y

# residual
Z = X - X_hat

# ----------------------------------------------------------
# 3. Prepare Curves
# ----------------------------------------------------------

# observation axis
y_axis = np.linspace(-10, 10, 500)

# conditional mean line
conditional_mean = alpha * y_axis

# confidence band
upper_band = conditional_mean + 2*np.sqrt(sigma_cond2)
lower_band = conditional_mean - 2*np.sqrt(sigma_cond2)

# ----------------------------------------------------------
# 4. Create Figure
# ----------------------------------------------------------
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=(
        "Conditional Mean and Confidence Tube",
        "Distribution of Observation Y",
        "Residual Error Structure",
        "Shrinkage Effect of Conditional Expectation"
    ),
    vertical_spacing=0.12,
    horizontal_spacing=0.10
)

# ==========================================================
# PANEL 1
# Conditional Mean + Confidence Region
# ==========================================================

# actual realizations
fig.add_trace(
    go.Scatter(
        x=Y,
        y=X,
        mode='markers',
        marker=dict(
            size=3,
            opacity=0.18
        ),
        name='True Samples (Y,X)'
    ),
    row=1,
    col=1
)

# conditional expectation line
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=conditional_mean,
        mode='lines',
        line=dict(width=4),
        name='Conditional Mean  E[X|Y]'
    ),
    row=1,
    col=1
)

# upper confidence
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=upper_band,
        mode='lines',
        line=dict(width=1, dash='dot'),
        showlegend=False
    ),
    row=1,
    col=1
)

# lower confidence
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=lower_band,
        mode='lines',
        fill='tonexty',
        line=dict(width=1, dash='dot'),
        name='95% Conditional Region'
    ),
    row=1,
    col=1
)

# ==========================================================
# PANEL 2
# Distribution of Y
# ==========================================================

hist_y, bins_y = np.histogram(Y, bins=50, density=True)

bin_centers = 0.5 * (bins_y[:-1] + bins_y[1:])

fig.add_trace(
    go.Bar(
        x=bin_centers,
        y=hist_y,
        opacity=0.6,
        name='Simulated Y Density'
    ),
    row=1,
    col=2
)

# theoretical density
pdf_y = norm.pdf(y_axis, 0, np.sqrt(sigma_y2))

fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=pdf_y,
        mode='lines',
        line=dict(width=4),
        name='Theoretical N(0,5)'
    ),
    row=1,
    col=2
)

# ==========================================================
# PANEL 3
# Residual Error Structure
# ==========================================================

fig.add_trace(
    go.Hexbin(
        x=Y,
        y=Z,
        nbinsx=35,
        name='Residual Density'
    ) if hasattr(go, "Hexbin") else go.Scatter(
        x=Y,
        y=Z,
        mode='markers',
        marker=dict(size=3, opacity=0.15),
        name='Residual Samples'
    ),
    row=2,
    col=1
)

# zero line
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=np.zeros_like(y_axis),
        mode='lines',
        line=dict(width=3, dash='dash'),
        name='Zero Correlation Line'
    ),
    row=2,
    col=1
)

# ==========================================================
# PANEL 4
# Shrinkage Visualization
# ==========================================================

# naive estimate
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=y_axis,
        mode='lines',
        line=dict(width=3, dash='dash'),
        name='Naive Estimate X=Y'
    ),
    row=2,
    col=2
)

# optimal estimate
fig.add_trace(
    go.Scatter(
        x=y_axis,
        y=alpha * y_axis,
        mode='lines',
        line=dict(width=4),
        name='Optimal Estimate'
    ),
    row=2,
    col=2
)

# sample paths
sample_indices = np.random.choice(np.arange(N), 50, replace=False)

for idx in sample_indices:
    fig.add_trace(
        go.Scatter(
            x=[Y[idx], Y[idx]],
            y=[Y[idx], X_hat[idx]],
            mode='lines',
            line=dict(width=1),
            opacity=0.2,
            showlegend=False
        ),
        row=2,
        col=2
    )

# ==========================================================
# 5. Layout
# ==========================================================

fig.update_layout(
    title=(
        "Advanced Gaussian Signal Estimation in Spindle Diagnostics"
        "<br><sup>Conditional Expectation, Residual Orthogonality, "
        "and Noise-Induced Shrinkage</sup>"
    ),
    template='plotly_white',
    height=950,
    width=1350
)

# ----------------------------------------------------------
# Axis Labels
# ----------------------------------------------------------

fig.update_xaxes(title_text="Observed Power Y", row=1, col=1)
fig.update_yaxes(title_text="True Temperature X", row=1, col=1)

fig.update_xaxes(title_text="Observed Power Y", row=1, col=2)
fig.update_yaxes(title_text="Density", row=1, col=2)

fig.update_xaxes(title_text="Observed Power Y", row=2, col=1)
fig.update_yaxes(title_text="Residual Error Z", row=2, col=1)

fig.update_xaxes(title_text="Observation Y", row=2, col=2)
fig.update_yaxes(title_text="Estimated Temperature", row=2, col=2)

# ----------------------------------------------------------
# 6. Show Figure
# ----------------------------------------------------------

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal

# ============================================================
# Q3 : Continuous Fault Monitoring of a Spindle Motor
# Visualization of Conditional Expectation and Residual Error
# ============================================================

# -----------------------------
# 1. System Parameters
# -----------------------------
# Hidden temperature
mu_x = 0
var_x = 1

# Mechanical noise
mu_q = 0
var_q = 4

# Observed spindle power
mu_y = 0
var_y = var_x + var_q

# Covariance
cov_xy = var_x

# Optimal MMSE gain
gain = cov_xy / var_y   # = 1/5 = 0.2

# Conditional variance
cond_var = var_x - (cov_xy**2 / var_y)

# -----------------------------
# 2. Simulated Measurements
# -----------------------------
np.random.seed(7)

N = 1200

# Generate temperature and noise
X = np.random.normal(mu_x, np.sqrt(var_x), N)
Q = np.random.normal(mu_q, np.sqrt(var_q), N)

# Sensor observation
Y = X + Q

# Optimal estimate E[X|Y]
X_est = gain * Y

# Residual error
R = X - X_est

# -----------------------------
# 3. Joint Gaussian Density
# -----------------------------
x_vals = np.linspace(-4, 4, 180)
y_vals = np.linspace(-9, 9, 180)

Xg, Yg = np.meshgrid(x_vals, y_vals)

pos = np.dstack((Xg, Yg))

cov_matrix = [
    [var_x, cov_xy],
    [cov_xy, var_y]
]

density = multivariate_normal(
    mean=[mu_x, mu_y],
    cov=cov_matrix
).pdf(pos)

# -----------------------------
# 4. Create Figure
# -----------------------------
fig = make_subplots(
    rows=1,
    cols=2,
    subplot_titles=(
        "Joint Gaussian Structure of Temperature and Power",
        "Residual Error After Conditional Estimation"
    ),
    horizontal_spacing=0.12
)

# ============================================================
# LEFT PANEL
# ============================================================

# Density contours
fig.add_trace(
    go.Contour(
        x=x_vals,
        y=y_vals,
        z=density,
        colorscale="Viridis",
        opacity=0.75,
        showscale=False,
        contours=dict(showlines=False),
        hoverinfo="skip"
    ),
    row=1,
    col=1
)

# Scatter cloud
fig.add_trace(
    go.Scatter(
        x=X,
        y=Y,
        mode='markers',
        marker=dict(
            size=3,
            opacity=0.22
        ),
        name='Observed Samples'
    ),
    row=1,
    col=1
)

# Optimal estimator line
y_line = np.linspace(-9, 9, 100)

fig.add_trace(
    go.Scatter(
        x=gain * y_line,
        y=y_line,
        mode='lines',
        line=dict(width=4),
        name='Optimal Estimate  E[X|Y]=0.2Y'
    ),
    row=1,
    col=1
)

# Naive estimator
fig.add_trace(
    go.Scatter(
        x=y_line,
        y=y_line,
        mode='lines',
        line=dict(width=2, dash='dash'),
        name='Naive Estimate  X=Y'
    ),
    row=1,
    col=1
)

# ============================================================
# RIGHT PANEL
# ============================================================

# Residual scatter
fig.add_trace(
    go.Scatter(
        x=Y,
        y=R,
        mode='markers',
        marker=dict(
            size=4,
            opacity=0.35
        ),
        name='Residual Error'
    ),
    row=1,
    col=2
)

# Zero correlation reference
fig.add_trace(
    go.Scatter(
        x=np.linspace(-9, 9, 100),
        y=np.zeros(100),
        mode='lines',
        line=dict(width=3),
        name='Orthogonality Line'
    ),
    row=1,
    col=2
)

# ============================================================
# 5. Layout Styling
# ============================================================

fig.update_layout(
    title=(
        "Continuous Fault Monitoring of a CNC Spindle Motor"
        "<br><sup>Conditional Expectation, Gaussian Geometry, "
        "and Orthogonality Principle</sup>"
    ),
    template="plotly_white",
    height=650,
    width=1250,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5
    )
)

# Axis labels
fig.update_xaxes(
    title_text="Hidden Temperature  X",
    row=1,
    col=1
)

fig.update_yaxes(
    title_text="Observed Spindle Power  Y",
    row=1,
    col=1
)

fig.update_xaxes(
    title_text="Observed Power  Y",
    row=1,
    col=2
)

fig.update_yaxes(
    title_text="Residual Error  Z = X - E[X|Y]",
    row=1,
    col=2
)

# -----------------------------
# 6. Display Figure
# -----------------------------
fig.show()
```

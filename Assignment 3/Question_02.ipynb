# Answer for Q2

Let

$$\mathbf{Z}=\begin{bmatrix} \mathbf{X}\\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$$

be jointly Gaussian with mean vector

$$\boldsymbol{\mu}=\begin{bmatrix} \boldsymbol{\mu}_X\\ \boldsymbol{\mu}_Y \end{bmatrix}$$

and covariance matrix

$$\boldsymbol{\Sigma}=\begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY}\\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}.$$‌

The joint density is

$$f_{\mathbf{X},\mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}} \exp\left( -\frac12 (\mathbf{z}-\boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z}-\boldsymbol{\mu}) \right).$$‌

We derive the conditional density

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}).$$



## Step 1: Definition of Conditional Density

By definition,

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}) = \frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}.$$

Since the marginal distribution of a multivariate Gaussian is also Gaussian,

$$\mathbf{Y}\sim \mathcal N(\boldsymbol{\mu}_Y,\boldsymbol{\Sigma}_{YY}),$$

and therefore

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac12 (\mathbf{y}-\boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_Y) \right).$$‌



## Step 2: Expand the Joint Quadratic Form

Let

$$\mathbf{u}=\mathbf{x}-\boldsymbol{\mu}_X, \qquad \mathbf{v}=\mathbf{y}-\boldsymbol{\mu}_Y.$$

Then

$$\mathbf{z}-\boldsymbol{\mu} = \begin{bmatrix} \mathbf{u}\\ \mathbf{v} \end{bmatrix}.$$‌

Using block matrix identities, the quadratic form in the exponent can be rewritten as

$$(\mathbf{z}-\boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z}-\boldsymbol{\mu})$$

$$ = (\mathbf{x}-\boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x}-\boldsymbol{\mu}^*) + (\mathbf{y}-\boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_Y),$$

where

$$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_Y),$$

and

$$\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX}.$$‌



## Step 3: Substitute into the Conditional Density

Substituting into

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}) = \frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})},$$

the terms involving only ($\mathbf{y}$) cancel.

Hence,

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n|\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac12 (\mathbf{x}-\boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x}-\boldsymbol{\mu}^*) \right).$$‌

Therefore,

$$\boxed{ f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n|\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac12 (\mathbf{x}-\boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x}-\boldsymbol{\mu}^*) \right) }$$

where

$$\boxed{ \boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_Y) }$$

and

$$\boxed{ \boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY} \boldsymbol{\Sigma}_{YY}^{-1} \boldsymbol{\Sigma}_{YX} }.$$‌

Thus, the conditional distribution of a jointly Gaussian random vector is again Gaussian.

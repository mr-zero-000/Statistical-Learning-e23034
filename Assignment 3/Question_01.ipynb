# Asnwer for Q1

## 1. Probability Space $(\Omega,\mathfrak F,P)$

The sample space is

$$ \Omega={\omega_1,\omega_2,\omega_3,\omega_4}, $$

where

$$ \omega_1:\text{ low-risk legitimate}, $$
$$ \omega_2:\text{ low-risk fraudulent}, $$
$$ \omega_3:\text{ high-risk legitimate}, $$
$$ \omega_4:\text{ high-risk fraudulent}. $$

Since $(\mathfrak F=\mathfrak P(\Omega))$, the $(\sigma)$-algebra is the power set of $(\Omega)$:

$$ \mathfrak F= {\emptyset,{\omega_1},{\omega_2},\ldots,\Omega}. $$

The probability measure is defined by

$$ P({\omega_1})=0.50, $$
$$ P({\omega_2})=0.10, $$
$$ P({\omega_3})=0.20, $$
$$ P({\omega_4})=0.20. $$

Also,

$$ 0.50+0.10+0.20+0.20=1, $$

hence $P$ is a valid probability measure.



# 2. Compute $\sigma(Y)$

The random variable $(Y)$ is defined by

$$ Y(\omega_1)=0,\qquad Y(\omega_2)=0, $$
$$ Y(\omega_3)=1,\qquad Y(\omega_4)=1. $$

Thus,

$$ {Y=0}={\omega_1,\omega_2}, $$
$$ {Y=1}={\omega_3,\omega_4}. $$

Therefore the $(\sigma)$-algebra generated by $(Y)$ is

$$ \sigma(Y)= { \emptyset, {\omega_1,\omega_2}, {\omega_3,\omega_4}, \Omega }. $$



# 3. Why $(E[X\mid Y])$ Must Be $\sigma(Y))$-Measurable

By definition, conditional expectation $(E[X\mid Y])$ must be measurable with respect to $(\sigma(Y))$.

Since the only nontrivial measurable sets in $(\sigma(Y))$ are

$$ {\omega_1,\omega_2} \quad \text{and} \quad {\omega_3,\omega_4}, $$

the function $(E[X\mid Y])$ cannot distinguish between points inside each set.

Hence $(E[X\mid Y])$ must be constant on:

$$ {\omega_1,\omega_2}, $$

and constant on

$$ {\omega_3,\omega_4}. $$

Let

$$ E[X\mid Y]= a\mathbb 1_{{Y=0}} +b\mathbb 1_{{Y=1}}. $$

We now determine $(a)$ and $(b)$.



# 4. Compute $(E[X\mid Y])$

## Case 1: $(Y=0)$

$$ a=E[X\mid Y=0]. $$

Using conditional probability,

$$ P(Y=0)=P(\omega_1)+P(\omega_2) =0.50+0.10 =0.60. $$

Now,

$$ E[X\mid Y=0] = 0\cdot P(X=0\mid Y=0) + 1\cdot P(X=1\mid Y=0). $$

But

$$ P(X=1\mid Y=0) = \frac{P(\omega_2)}{P(Y=0)} = \frac{0.10}{0.60} = \frac16. $$

Hence,

$$ a=\frac16. $$



## Case 2: $(Y=1)$

$$ P(Y=1)=0.20+0.20=0.40. $$

Then

$$ P(X=1\mid Y=1) = \frac{0.20}{0.40} = \frac12. $$

Thus,

$$ b=\frac12. $$

Therefore,

$$ E[X\mid Y] = \frac16\mathbb 1_{{Y=0}} + \frac12\mathbb 1_{{Y=1}}. $$



# 5. Verification of the Defining Property

We verify that for every $(A\in\sigma(Y))$,

$$ \int_A E[X\mid Y]dP = \int_A XdP. $$

The sets in $(\sigma(Y))$ are:

$$ \emptyset,\quad \Omega,\quad {\omega_1,\omega_2},\quad {\omega_3,\omega_4}. $$



## For $(A={\omega_1,\omega_2})$

Left side:

$$ \int_A E[X\mid Y]dP = \frac16 \times P(A) = \frac16\times0.60 = 0.10. $$

Right side:

$$ \int_A XdP = 0(0.50)+1(0.10) = 0.10. $$

Hence equal.


## For $(A={\omega_3,\omega_4})$

Left side:

$$ \int_A E[X\mid Y]dP = \frac12\times0.40 = 0.20. $$

Right side:

$$ \int_A XdP = 0(0.20)+1(0.20) = 0.20. $$

Hence equal.

Therefore the defining property holds.

# 8. Mean Square Error (MSE)

The optimal estimator based on $(Y)$ is

$$ \hat X=E[X\mid Y]. $$

Hence,

$$ \text{MSE} = E[(X-E[X|Y])^2]. $$

Using conditional variance,

$$ \text{MSE}=E[\operatorname{Var}(X|Y)]. $$



## Variance when $(Y=0)$

Since $(X)$ is Bernoulli with parameter $(1/6)$,

$$ \operatorname{Var}(X|Y=0) = \frac16\left(1-\frac16\right) = \frac{5}{36}. $$



## Variance when $(Y=1)$

$$ \operatorname{Var}(X|Y=1) = \frac12\left(1-\frac12\right) = \frac14. $$

Therefore,

$$ \text{MSE} = 0.60\left(\frac{5}{36}\right) + 0.40\left(\frac14\right). $$

$$ \frac{1}{12}+\frac{1}{10} = \frac{11}{60}. $$

Thus,

$$ \boxed{\text{MSE}=\frac{11}{60}\approx0.1833} $$



# 9. MSE Using Only the Global Average

The global estimator is

$$ E[X]=P(X=1)=0.10+0.20=0.30. $$

Thus the estimator is the constant $(0.30)$.

The MSE becomes

$$ E[(X-0.30)^2]. $$

Since $(X)$ is Bernoulli($(0.30))$,

$$ \operatorname{Var}(X)=0.30(0.70)=0.21. $$

Therefore,

$$ \boxed{\text{MSE}=0.21} $$


## Relation to Law of Total Variance

The Law of Total Variance states

$$ \operatorname{Var}(X) = E[\operatorname{Var}(X|Y)] + \operatorname{Var}(E[X|Y]). $$

We found:

$$ \operatorname{Var}(X)=0.21, $$

$$ E[\operatorname{Var}(X|Y)] = \frac{11}{60} \approx0.1833. $$

Hence,

$$ \operatorname{Var}(E[X|Y]) = 0.21-0.1833 = 0.0267. $$

This shows that using $(Y)$ reduces prediction error because part of the uncertainty is explained by the customer's risk class.



# 10. Entropy and Mutual Information

The joint distribution is


\
\begin{array}{c|c|c}
X & Y & P(X,Y)\\
\hline
0 & 0 & 0.50\\
1 & 0 & 0.10\\
0 & 1 & 0.20\\
1 & 1 & 0.20
\end{array}




# 10(a) Entropy $(H(X))$

$$ P(X=0)=0.70, \qquad P(X=1)=0.30. $$

Entropy:

$$ H(X) = -\sum_x p(x)\log_2 p(x). $$

Hence,

$$ H(X) = -0.70\log_2(0.70) -0.30\log_2(0.30). $$

$$ \boxed{H(X)\approx0.881\text{ bits}} $$



# 10(b) Conditional Entropy $(H(X|Y))$

$$ H(X|Y) = \sum_y P(y)H(X|Y=y). $$



## For $(Y=0)$

Distribution:

$$ \left(\frac56,\frac16\right). $$

Thus,

$$ H(X|Y=0) = -\frac56\log_2\frac56 -\frac16\log_2\frac16. $$

$$ \approx0.650. $$



## For $(Y=1)$

Distribution:

$$ \left(\frac12,\frac12\right). $$

Thus,

$$ H(X|Y=1)=1. $$

Therefore,

$$ H(X|Y) = 0.60(0.650)+0.40(1). $$

$$ \boxed{H(X|Y)\approx0.790\text{ bits}} $$



# 10(c) Mutual Information $(I(X;Y))$

$$ I(X;Y) = H(X)-H(X|Y). $$

Hence,

$$ I(X;Y) = 0.881-0.790. $$

$$ \boxed{I(X;Y)\approx0.091\text{ bits}} $$


# 10(d) Percentage of Uncertainty Removed

$$ \text{Percentage removed} = \frac{I(X;Y)}{H(X)}\times100\%. $$

Thus,

$$ \frac{0.091}{0.881}\times100\%. $$

$$ \boxed{\approx10.3\%} $$

Therefore, knowing the customer risk class removes approximately $(10\%)$ of the uncertainty regarding whether a transaction is fraudulent.
